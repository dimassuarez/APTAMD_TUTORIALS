## Installing Conda - JupyterLab on Windows 11

The `jupyter-lab` interactive notebook editor is an application or package of the [Python](https://www.python.org/) language. Currently, Python is probably the most widely used programming language by the scientific community for all kinds of tasks, from simple calculator-style applications to complex computational chemistry simulations or regressions using *machine learning* models. Therefore, before installing `jupyter-lab`, we must install a Python software distribution on our computer. For beginners, it is recommended to install a *minimal* Python distribution such as `Miniconda3`.  

 1 - Download the Miniconda3 Python environment installation program available at   [Miniconda3 install website.](https://docs.conda.io/projects/miniconda/en/latest/miniconda-install.html)

 2 - Run the installation program following the instructions on the Miniconda page, accepting the default options.
 
 [x]:![](instalacion_1.png)
<img src="./instalacion_1.png">

 3 - See basic options for starting and managing the conda environment described in the [user guide](https://docs.conda.io/projects/conda/en/latest/user-guide/getting-started.html). 

 4 -  Start the conda environment as described in the user guide. From the Windows Start menu, find and open the `Anaconda Prompt` program. A command window (*terminal*) for the Windows `PowerShell` interpreter will open, ready to accept commands from the Miniconda3 environment.
 
 [x]:![](prompt.png)
<img src="./prompt.png" width="600" height="400">

 5 - Install the basic Python packages for numerical computation and data analysis (Miniconda3 contains only a few [packages](https://pypi.org/) of programs and libraries available for the Python programming language). To do this, enter the following command in the terminal:

 ```powershell
(base) PS C:\Users\Student> conda install numpy matplotlib scipy
```

The installation program also selects other packages necessary for the selected options, asking you to confirm their installation. 

 [x]:![](instalacion_paquetes.png)
<img src="./instalacion_paquetes.png" width="700" height="400">

 6 - Once all packages have been downloaded and installed, the command window prompt will become active again. Then install the jupyter-lab application.

`(base) PS C:\Users\Student> conda install jupyterlab`

 7 - From the 'conda PS' command window, start the jupyter-lab server.

`(base) PS C:\Users\Student> jupyter-lab`

Windows may ask you which application you want to use with `jupyter-lab`. We recommend selecting (or installing) the [Firefox browser](https://www.mozilla.org/es-ES/firefox/new/). 

 [x]:![](windows_question.png)
<img src="windows_question.png"  width="300" height="300"> 

The command window remains locked while the jupyter-lab server is running: Do not close it!

  8 - Once the notebook editor has started, you will access its graphical environment through the browser. Among the available utilities, you can create computational notebooks using the Python language. To close the editor, select the *File->Shut Down* option.

  [x]:![](nb1.png)
 <img src="nb1.png"  width="640" height="344">

#### CONDA / JupyterLab on WSL

If you have installed a Linux environment on your computer using [WSL](https://learn.microsoft.com/en-us/windows/wsl/install) then the installation of CONDA/JupyterLab is feasible following the same instructions.

#### Python environments
In case that you plan to use other Python software then it is strongly recommended to manage different [Python environments](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html) for the software packages.
