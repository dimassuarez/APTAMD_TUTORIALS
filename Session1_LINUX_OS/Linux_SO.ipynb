### Example of a Code cell 

Small computer programs can be entered in  *code* type cells. Different computer languages can be used, but Python is the most used. Each code  cell can be executed individually by pressing `CTRL+Enter`. The notebook displays the *output* of the code below the executed cell.

The following cell is an example of Python code that performs a quadratic fitting of equilibrium constant values vs pressure. 

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Data
P=np.array([1,400,1000,1400,2000,2400,3000])
lnK=np.log ( np.array([ 1.746 , 2.079 , 2.657 , 3.072  , 3.773  , 4.311 ,  5.21 ])*1E-5 )
# Fitting eq
def lnK_fit(P,a,b,c):
    lnK= a+b*P+c*P*P  
    return lnK
# Fitting method
[a,b,c],pcov = curve_fit(lnK_fit,  P, lnK)
# Parameters
print ('Fitting equation')
print ( f'lnK = {a:f} + {b:f} * P + {c:e} * P**2 ')
# Plot 
plt.plot(P,lnK,'or',markersize=5)
Px=np.linspace(0,5000,100)
plt.plot(Px, lnK_fit(Px,a,b,c) ,'-m')
plt.grid()
plt.xlabel('P(bar)'); plt.ylabel('ln K')
plt.legend(['Exp', 'Fitting'])
plt.show()
````

## **Secure connection to remote Linux computers**

### SSH text-mode connection from Windows PowerShell

The [OpenSSH](https://www.openssh.org/) package provides various applications for secure communications (under encryption) between  computers. It is available both in Linux and in Windows-based OS.  

A first method to login to a remote Linux computer uses the ``ssh`` command. The ssh command is executed on the source (or local) machine and opens a session in text mode in the user's account on the remote machine. Under the Windows-OS, ``ssh`` is accessible at the command line of **PowerShell**.

Start [Powershell](https://learn.microsoft.com/en-us/powershell/scripting/windows-powershell/starting-windows-powershell) and then enter the following command:

### SSH-tunneled VNC connection to graphical desktops: PowerShell and TigerVNC

The desktop is the graphical interface between the user and the OS (applications, data files,  etc.). Follow these steps to connect to your desktop: 

* Ask the system administrator of the remote server if the desktop service has been configured for you.

* A client/server application running on the local computer is needed to launch and interact with graphical applications on remote machines. This is an executable binary file that does not require further installation.
  
The VNC connection is a two-step (double authentication, SSH + VNC) process:

1. An **SSH tunnel** between the local computer and the remote server must be open for safe (encrypted) communication. To this end, open a Powershell window and enter the following `ssh` command:

```bash
ssh -L 60X:127.0.0.1:59X aptauser@computer_address -N 
```
where X is a two-digit personal ID number assigned to your desktop by the system administrator. *SSH authentication* will be requested.

**To keep the SSH tunnel open, keep the Powershell window open!**

2. Start the VNC viewer program in the local computer. Connect the VNC viewer to `localhost:60X` and then  enter your *VNC authentication* password.

The Linux OS has a very powerful operator known as    *pipe* that  is represented by the vertical bar `|` . This operator couples the output of one program or application with the input of another as represented in the following sequence:
```bash
 prog1 < input | prog2 | prog3 ... | progN > output
```
so that the output of `prog1`is piped as input into `prog2` whose output is, in turn, piped to `prog3` and so on.

## **About BASH  scripts**

Bash scripts are files containing Linux commands and some elementary control-flow constructs  that determine how instructions are executed. 

Some common features and programming structures in BASH scripts are :

* Scripts start with the shebang `#!` followed by the path to Bash, usually `/bin/bash`.
* The `#` symbol precedes explanatory comments.
* Scripts define and use variables that store data.  The `=` symbol assigns values to variables ( = sign without spaces).
```bash
#!/bin/bash
name="World"         # Assign a value to a variable
echo "Hello, $name!" # To access the value of a variable, prefix it with a dollar sign
```
* The OS declares special  environment variables (`$HOME, $USER, $PWD, $PATH` ...) that can affect how programs and scripts are run.
* `if...else..fi` statements provide a way to execute one block of code if a condition is true and another block if it is false.
```bash
if [ "$SIEVE" -le 1 ]; then  # Checking input value for the SIEVE variable 
    echo "SIEVE must be > 1"
    exit 1 
else
    echo "MD frames processed with SIEVE=$SIEVE"
fi
```
* `for` loops *iterate* over a list of items or a range of numbers. They are useful for repeating tasks a specific number of times.
```bash
for file in $(ls *.pdb)  # The loop runs over a list of pdb filenames
do
    NAT=$(grep -c 'ATOM  ' $file)  # $( command )  captures the output of Linux commands
    echo "PDB $file contains $NAT ATOM records"
done
```

## **Job control**

Molecular modeling tasks on Linux computers demand computational resources in the form of CPUs, GPUs, RAM space, hard-disk space, etc.  Some tasks can be performed by running a software application interactively as, for example, structure visualization and manipulation using programs like ChimeraX. Other tasks related with molecular simulation and data analysis are normally performed by means of auxiliary script files, that implement a workflow combining various programs and/or data sets. In this case, the script defines a *job* that, depending on the job type  and the available computer resources, may require  a long *wall time*  to be completed, ranging from seconds to hours, days, weeks, ... 

The main job execution modes are:

### **Interactive (or foreground) mode** 

The job script `job.sh` is run at the command-line interface likewise other Linux OS commands.
```bash
./job.sh 
```
Note that the command-line shell can not process any other command as long as the job is running. Interactive job execution can be interrupted using `CTRL+C` keyboard shortcut.

### **Background mode**  
The job script or other commands can be run as a background process or service. For example, an interactive job can be suspended pressing `CTRL+Z` and resumed either in foreground or background mode with the `fg` or `bg` commands, respectively.

* **Q12** Reproduce the sequence of command-line actions that control the execution of the `sleep` command. Note that `%1` is a job id signaling the *first* job attached to the terminal. 

```bash 
[aptauser@server aptauser]$ sleep 60
^Z
[1]+  Stopped                 sleep 60
[aptauser@server aptauser]$ fg %1
sleep 60
^Z
[1]+  Stopped                 sleep 60
[aptauser@server aptauser]$ bg %1
[1]+ sleep 60 &
[aptauser@server aptauser]$ jobs
[1]+  Running                 sleep 60 &
[aptauser@server aptauser]$ kill %1
[1]+  Terminated              sleep 60

```


Any job can be started directly on the background mode using the `&` operator as illustrated in the following example:     
``` bash
[aptauser@server aptauser]$ sleep 60 &
[1] 3724961
[aptauser@server aptauser]$ jobs
[1]+  Running                 sleep 60 &
```   
The background job is now also  identified with a pid number (process id). While running, the process is linked to the terminal session so that closing the terminal or logging out from the computer would kill the job. To detach the backrground job from the command-line terminal, we employ the command `nohup` 
```bash
aptauser@server aptauser]$ nohup sleep 60 &
[1] 3725798
nohup: ignoring input and appending output to 'nohup.out
```
When jobs are started on the background mode, the standard output should be redirected to text file. By default the `nohup` command redirects standard output to the `nohup.out` file, but it can be convenient to select another file using the `>` operator.

### **Job control by queue systems**

When multiple users have access to a common computer infrastructure and have much appetite for running jobs,  resource management software is needed to  control jobs and prevent system overload. Usually, this  software creates a *queue system* to which users must submit their jobs. Then the software  manages, schedules, and prioritizes submitted jobs, ensuring efficient resource utilization. Besides submitting jobs, users are allowed to  monitor, hold, and cancel jobs. Queue systems are typically used in computing clusters so that jobs are  distributed over the compute nodes. 

In a Linux cluster, all jobs running molecular dynamics (MD) simulations or analysis should be submitted by means of a queue system because they demand much computational resources. In the cluster, users initiate their sessions on the front-end server (`computer_address`) from which jobs are submitted to different compute nodes.
