## **Structural elements of DNA aptamers** 

IUPAC defines DNA aptamers  as single stranded (ss)  oligonucleotides originating from in vitro selection which, starting from random sequence libraries, optimize the composition for high-affinity binding to a given target. As you know, nucleotides are the building blocks of aptamers and other nucleic acids. They consist of a sugar molecule (deoxyribose in DNA) bonded to a negatively charged phosphate group and a nitrogenous base (adenine (A), cytosine (C), guanine (G) and thymine (T) in  natural DNA).

### **Backbone conformation of ssDNAs**

As expected, aptamers can adopt multiple conformations through   internal rotation motions around the covalent bonds that constitute the *backbone* of the polymer.  Thus, the conformational state of a given $\ce{C-C}$ rotatable bond is characterized in terms of the **torsion angle**  between two intersecting planes formed by four consecutively bonded atoms $\ce{X-C-C-Y}$, where $\ce{X, Y}$ are two terminal atoms attached to the central bond. 

[x]:![](PNG/dihedral.png)
<img src="./PNG/dihedral.png" width="320" height="160" style="display: block; margin: 0 auto">

**Fig 4** $\ce{X-C-C-Y}$ torsion angle definition.

For biomolecules like oligonucleotides and peptides, the torsion angles around all the backbone bonds and  selected side chain bonds are denoted with special symbols (see Fig. 2). For example, the $\alpha$ torsion angles in oligonucleotides are defined by the $\mathrm{O3'_{(i-1)}-P_{i}-O5'_{i}-C5'_{i}}$ atoms of consecutive nucleotides , $\beta$ is defined by  $\mathrm{P_{i}-O5'_{i}-C5'_{i}-C4'_{i}}$, while the $\chi_{i}$  angle refers to the bond that links the base and the sugar and adopts either *syn* and *anti* forms ([more about backbone conformation](https://nakb.org/basics/nucleotides.html#BACKBONE/)). To grasp the actual conformational freedom of oligonucleotides, it may be interesting to note that  up to 96  [conformational classes](https://nakb.org/basics/nucleotides.html#CONFORMERS/) at the  dinucleotide level have been defined basing on  statistical analysis of experimental data .

### **Secondary structure: 2D motifs**

Fortunately, the majority of functional biomolecules do not behave as random coils. Thus, the backbone torsion angles are flexible but restricted to sterically allowable regions. Moreover,  non-covalent intramolecular interactions as well as electrostatic effects (e.g. interaction with solvent and counterions) largely determine the spatial folding of aptamers, which tend to adopt characteristic conformation(s). A basic knowledge of the main interactions and the most common structural motifs is therefore required to understand the structure of aptamers.     

Non-covalent attractive interactions give rise to two major type of intra-strand contacts:  

* **base stacking** involving $\pi \cdot\cdot\cdot \pi$ contacts  that allow nucleotides to pack densely.  

* **base pairing** involving  hydrogen-bond  contacts between nucleobases. In the well-known **canonical Watson-Crick base pairing**,   adenine binds to thymine ($\ce{A\cdot T }$) with two H-bonds and cytosine to guanine ($\ce{C\cdot G }$) with three H-bonds. But most aptamers  incorporate also  **non-canonical base pairings** and **mismatches** like the *Hoogsteen* base pair where a purine base ($\ce{A}$  or $\ce{G}$ ) flips $180^\circ$ into a *syn* conformation that changes the identity of the H-bonds with the corresponding pyrimidine base, the *Wobble* base pair where the displacement of one of the bases accommodates mismatches like $\ce { G\cdot T}$, and the *Sheared* base pair like the $\ce { G\cdot A}$ pairing “side-by-side”.  Visit the [NAKB portal](https://nakb.org/basics/basepairs.html) or the [Jena-library](https://jenalib.leibniz-fli.de/IMAGE_BPDIR.html) for examples and further information. 

The spatial conformations  of an aptamer molecule can be first described in terms of its **secondary structure (2D)**, which shows the local folding patterns formed by the intra-strand base-stacking and base-pairing contacts.  More particularly, base pairing within the aptamer chain results in different 2D structural motifs including:     

* **Single-stranded segments**. 
* **Helices**, constituted by base-stacking interactions, regardless of base-pairing  and backbone connectivity. A helix is composed of at least two stacked pairs.
* **Stems** ,  as a helix consisting of canonical pairs (WC or wobble) without backbone breaks. 
* **Hairpins**, a double-stranded stem topped by a single-stranded loop. It is the most common motif.
* **Internal loops**, flexible regions of unpaired bases within a stem that are prone to act as binding sites.
* **Bulges**, extra bases on one of the strands of a stem that cause distortion and introduce flexibility.
* **Multi-branched junctions/loops**, junction points/loops where three or more stems meet. The central loop usually results in a flexible binding site.

## **Role of hydration and ion interactions**

A nucleic acid molecule with $N$ nucleotides presents an overall negative charge of $(N-1)$ distributed evenly across the phosphate groups of its backbone. Then it is not a surprise that both  water molecules and positive counterions ($\ce{Na+, K+, Mg^2+, ...}$) behave as  active structural components that influence the folding of DNA/RNAs aptamers based on the following effects:   
 
* **Hydrophobic packaging**  Nucleobases are relatively *hydrophobic*  and therefore, they tend to form $\pi$-stacking clusters as in helices and stems, avoiding thus to disrupt water H-bond networks.

* **Formation of multiple hydration layers**  Direct H-bond interactions with polar groups give rise to a *structured* first hydration shell of $\simeq 15-25$ water molecules per nucleotide with residence times ranging from ps to ns, bridging thus backbone and side-chain atoms and resulting in local hydration patterns. Similarly, water molecules can also play a structural role in aptamer binding sites. 

* **Electrostatic shielding** Water, as a continuum solvent, has a high dielectric constant (~80). In addition, counterions tend to cluster around ionic solute molecules of opposite charge. The two effects are taken into account by the Debye-Hückel theory to predict a screening factor $\exp \left(-\kappa r \right) $ by which the ordinary Coulomb  electrostatic interactions are reduced by the presence of solvent and counterions. The Debye length $1 / \kappa$ amounts to $\sim 8 \rm{\mathring{A}}$ at physiological ionic strength conditions so that electrostatic repulsion between nearby phosphate groups located at $\sim 10 \rm{\mathring{A}}$ are effectively dumped. 

As a result of screened Coulomb effects, the association of counterions with nucleic acids and aptamers spreads over a broad range, from very loose associations to tight binding, and plays a key role in the structure and dynamics of nucleic acids. For example, monovalent ions help stabilize the Watson-Crick base pairs in the helices (stems). Divalent cations tend to bind more strongly and in a more sequence-specific manner than univalent ions with the groove regions and the phosphate backbone atoms, being able to stabilize tight twists and complex junctions between local structural elements. Of particular interest can be also the G-quadruplex motif (see below), which is associated with the presence of $\ce{K+}$ ions.

The determination of precise counterion and hydration patterns is still challenging for both experimental and computational techniques. Nonetheless, it must be emphasized that **a careful choice of  salt concentration conditions is required for the molecular modeling of aptamers**.     




## **2D model prediction**

Aptamers recognize their target through structural rather than sequence motifs. However, obtaining a reliable 2D structure from a given 1D sequence is a prerequisite for predicting realistic 3D structures. In principle, motifs such us hairpins, loops, bulges and junctions can be predicted from 1D sequences using a large number of bioinformatic methodologies. The 2D structures of both ssDNA and RNA molecules can be considered by these tools, the RNA structures exhibiting a larger variety of 2D motifs.  Depending on their methodological approach, these tools  are  classified as: 

* **Statistical thermodynamic methods**, aiming to estimate the free energy of the folded structures combining statistical thermodynamic models of molecular zippers with accurate experimental data and/or empirical parameters. 
* **Comparative sequence analysis**, traditional bioinformatic approach based on sequence alignment algorithms.  
* **Machine / deep learning methods**, trained to reproduce complex correlation patterns between sequence and 2D structures. 
* **Hybrid approaches**, using a weighted scoring function combining various methods.

###  **Thermodynamic 2D methods are still the methods of choice : `mfold`** 
Basing on our previous computational experience, we employ preferentially thermodynamic methods such as that implemented in the popular [`mfold` software]( https://doi.org/10.1093/nar/gkg595), which yields 2D models ranked in terms of $\Delta G_{fold}$ values. Starting at the 1D sequence, `mfold` builds a series of 2D structures combining loop and stack motifs and estimates their  $\Delta G_{fold}$ using high-quality experimental thermodynamic data for canonical base paring including nearest neighbor effects. The reliability of thermodynamic methods has been noticed in a recent validation study by [Binet at al](https://doi.org/10.1186/s12859-023-05532-5). In this work,  several prediction tools were applied to a dataset containing single-stranded nucleic acids with hairpins, pseudoknots, and multiple-way junctions, concluding that `mfold` provides the most accurate predictions in DNA sequences!

#### **Running `mfold` as a web service**
The [mfold](https://www.unafold.org/mfold/applications/dna-folding-form.php) method is accessible at a web server. For a typical aptamer 2D prediction, users should provide at least the following arguments:
* Sequence name.
* The ordered list of nucleotides using the one-letter FASTA format (A, T, C, G).
* The folding temperature in ºC. 
* The ionic conditions corresponding to the concentration (M or mM) of Na+ and/or Mg2+. 

Other options in the web service are usually left to their default values.

### **Using other 2D modeling tools: `NUPACK`**

Certainly, the `mfold` code is a well tried and tested tool, but it is also  a *legacy* application that is not actively maintained. Therefore, it is  advisable to compare `mfold` results with those  provided by other algorithms and/or tools like the `NUPACK` software suite. This program, which also implements statistical thermodynamics algorithms, has been written as a Python module, is under active development, and allows users to predict secondary structures including pseudo-knots for an arbitrary number of interacting nucleic acid strands. It serves also as a design tool searching for sequences with desired base-pairing properties. `NUPACK` is also highly recommendable because of the extensive online [documentation](https://docs.nupack.org) about  the implemented algorithms and the modes of usage.  

Much of the `NUPACK` software can be conveniently run using the web application at  [www.nupack.org](https://www.nupack.org/) (an annual subscription fee is required for running the web service). Alternatively, you can register at the website and  download the Python code to  run `NUPACK` locally. 

The `NUPACK` application is normally installed within a  `conda environment`, which is a kind of software container that bundles  the main application with all software dependencies.  To activate the `NUPACK` environment , we use another [wrapper script](SCRIPTS/nupack_env.sh) :

* **Q9** Copy the following Python script  into a text file with extension `.py` (e.g., `Glu4.py`). Edit the script and adapt it to your needs. 

````python
#  NUPACK PYTHON SCRIPT TO PREDICT SINGLE STRAIN 2D MODELS
from nupack import *

# Auxiliary function to print dbn files
def dbnfile(filename,info,seq,struc):
    with open(filename, 'w') as f:
        print(info,file=f)
        print(seq,file=f)
        print(struc,file=f)
    f.close()
    return

# 1. Model Configuration. Define here T and salt concentration
T=20             # oC
NA_CONC=0.0      # M
MG_CONC=0.0      # M
ENERGY_GAP=0.5   # kcal/mol (suboptimal structures)
my_model = Model(material='DNA', celsius=T, sodium=NA_CONC, magnesium=MG_CONC)
# 2. Define here the aptamer sequence  .
aptamer_seq = ''      # 1-letter fasta sequce
name_seq= ''          # id name 
my_strand = Strand(aptamer_seq, name= name_seq )

# Calculations start here
print('NUPACK MODEL')
print('=============')
print(f">1D structure (input sequence) ")
print(aptamer_seq)
print(f">1D structure (rna-equiv) ")
print(aptamer_seq.replace("T","U").replace("t","u"))
print(f"Conditions:  T={T} oC, NA_CONC={NA_CONC} (M), MG_CONC={MG_CONC} (M)")
# 3. Compute Minimum Free Energy structures with the mfe function 
mfe_structures = mfe(strands=[my_strand], model=my_model)
# 4. Compute set of suboptimal structures with a stacking state within a specified energy gap
subopt_structures = subopt(strands=[my_strand], energy_gap=ENERGY_GAP, model=my_model)
# 5. Print info for MFE structures
print('MINIMUM FREE ENERGY STRUCTURES}')
print('===============================')
n=1
for solution in  mfe_structures :
    print(f"2D struture {n} {solution.structure} ")
    print(f"   Free Energy     : {solution.energy:.2f} kcal/mol")
    print(f"   Stack Energy    : {solution.stack_energy:.2f} kcal/mol")
    mfe=solution.energy
    s2D=solution.structure.dotparensplus()
    dbnfile(f'{name_seq}_{n}.dbn',f'>Model {n} Free Energy = {mfe:.2f} kcal/mol', aptamer_seq, s2D)
    n=n+1
# 6. Print info for suboptimal structures
print('SUBOPTIMAL STRUCTURES}')
print('======================')
# Print info other suboptimal structures within the ENERGY_GAP  
dG0=mfe
s2D0=s2D
for solution in subopt_structures:
    dG=solution.energy
    s2D=solution.structure.dotparensplus()
    if  ( abs(dG - dG0) > 0.001 ) &  ( s2D != s2D0 ) :
        print(f"2D struture {n} {solution.structure} ")
        print(f"   Free Energy     : {solution.energy:.2f} kcal/mol")
        print(f"   Stack Energy    : {solution.stack_energy:.2f} kcal/mol")
        dbnfile(f'{name_seq}_{n}.dbn',f'>Model {n} Free Energy = {dG:.2f} kcal/mol', aptamer_seq, s2D)
        n=n+1
    dG0=dG
    s2D0=s2D
````

## **3D model prediction**

A quick survey of the PDB database points out that the number of experimentally solved RNA structures is far superior to that of DNA structures. This fact may be related to the relatively low flexibility of RNA molecules in spite of their tendency to fold in complicated forms. Hence, it seems reasonable that most  of the bioinformatic tools that provide 3D models of nucleic acids  have been specifically designed, parameterized  and validated  for RNA. Nonetheless, they are also useful for modeling DNA aptamers since the prevailing strategy  repurposes RNA-specific algorithms by following these steps: 

1. DNA sequences are converted into RNA ones.
2. From the 1D(RNA) and 2D(DNA) structures,  use RNA prediction tools such as [`RNAComposer`](https://rnacomposer.cs.put.poznan.pl/) or [`rna_de_novo`](https://docs.rosettacommons.org/docs/latest/application_documentation/rna/rna-denovo).
3. RNA 3D models are transformed back into DNA using molecular modeling software.

Going from 2D to 3D is the key step in this protocol while the RNA to DNA transformation can be easily automated using the APTAMD tools.    

###  **`RNAComposer` : *fast* and user-friendly**

[`RNAComposer`](https://rnacomposer.cs.put.poznan.pl/) is a quite popular web server that implements a *fast* bottom-up algorithm for RNA modeling. It decomposes first the input 2D structure into the different motifs (loop, stems, etc.), selects geometries for the 2D elements from a structural database of RNA fragments,  uses a translation algorithm to generate coordinates for unmatched motifs, and builds a global 3D model by performing superimposition operations and limited structural relaxation.

From previous computational experience, the `RNAComposer` results can be considered as reliable *initial* structures for aptamer sequences exhibiting  relatively simple 2D structures. The web server, which is quite user-friendly, works either in interactive or in batch modes. ###  **`rna_de_novo`: computationally *expensive* and less user-friendly**

Clearly, `RNAComposer` is a *fast* interactive tool for predicting 3D RNA structures by assembling multiple fragments. Computational speed is achieved by means of a quick exploration of the conformational space for relative orientations among RNA fragments. However, due to such limited sampling,  the resulting tertiary structure may miss important details especially in the case of complex 2D structures. For this reason, it is worth considering other tools like the `rna_de_novo` program, which is included in the [Rosetta software suite](https://rosettacommons.org) for biomolecular modeling.      

The [`rna_de_novo`](https://docs.rosettacommons.org/docs/latest/application_documentation/rna/rna-denovo) tool implements the [FARFAR2 algorithm](https://docs.rosettacommons.org/docs/latest/FARFAR2) for RNA structure prediction. Likewise RNAComposer, FARFAR2 relies on experimental ( NMR and X-ray) data of RNA fragments. But FARFAR2  incorporates  more powerful sampling methods as those typically used for predicting the structure of biomolecular complexes (i.e., *docking* calculations). In this way,  `rna_de_novo` is able to build *thousands* of structurally-feasible 3D models that are subsequently relaxed and  scored  using a customized force field and classified by means of clustering techniques. Of course, the price to pay is a higher computational cost so that `rna_de_novo` jobs require wall times ranging from hours to days depending on the computer resources.

The  [Rosie](https://rosie.graylab.jhu.edu/) web server offers online access to all the Rosetta tools including  `rna_de_novo`. Thus, registered users can specify and submit `rna_de_novo` jobs to a queue system, but job execution can be delayed a long time. Therefore, installing `rna_de_novo` on  local computers seems a better option. On a Linux  cluster, `rna_de_novo` jobs should be always submitted to the queue system. Each job is defined by a BASH script.  

* **Q13** Copy and paste the `rna_de_novo` script below into a `.sh` file. Examine carefully the script.   Which statements should you modify to run a `rna_de_novo`job? Which are the output files to be generated?
    * **`rna_de_novo` jobs might be submitted  at the end of the session**

````python
#!/bin/bash
# Driver script for 3D RNA prediction using rna_de_novo
# Input
SEQ=''   # Define here the DNA (or RNA) sequence (one-letter code)

STRUC_2D='' # Define here the 2D structure in dot-bracket notation 

NAME=''   # Define a short name (output file prefix)  

# As rna_de_novo only accepts rna sequence in lower case characters
# the following line fixes input sequence if needed
SEQ=$(echo $SEQ | tr '[:upper:]' '[:lower:]' | sed 's/t/u/g')
echo "RNA sequence= $SEQ "

# Parameters controlling rna_de_novo execution
OUTPUT="${NAME}"       # Output file prefix 
NCORES=8          # Number of computer cores to be used
NSTRUCT=500       # Number of 3D models to be generated
RMSD_THRES=6.0    # RMSD threshold (in Ang) for clustering 3D models 

# BASH environment variables (To be adapted for your environment)
export ROSETTA_BIN_DIR="/opt/apps/AL9/rosetta.source.release-371/main/source/bin"
export ROSSETTA_RELEASE="linuxgccrelease"
export PATH=/usr/lib64/openmpi/bin:$PATH
export LD_LIBRARY_PATH=/usr/lib64/openmpi/lib:/usr/lib64

# Only rna_de_novo uses MPI parallelism efficiently!
mpirun -np ${NCORES}  $ROSETTA_BIN_DIR/rna_denovo.mpi.${ROSSETTA_RELEASE} -sequence ${SEQ} \
       -secstruct ${STRUC_2D} -nstruct ${NSTRUCT} -out:file:silent  ${OUTPUT}_dock.out -minimize_rna

# RMSD Clustering of the 3D structures. Clusters are collected in the clusters.out file. 
$ROSETTA_BIN_DIR/rna_cluster.${ROSSETTA_RELEASE} -in:file:silent ${OUTPUT}_dock.out  -out:file:silent ${OUTPUT}_cluster.out -cluster:radius ${RMSD_THRES}

# Extract cluster representatives ranked in terms of their score energy
$ROSETTA_BIN_DIR/rna_extract.${ROSSETTA_RELEASE}  -in:file:silent ${OUTPUT}_cluster.out -in:file:silent_struct_type  rna

# Extract the clusters ID,  scoring energy and population  
echo "# Repr. ID,  Scoring, Population" > ${OUTPUT}_summary.dat
grep "SCORE" ${OUTPUT}_cluster.out | sed '1,1d' |  awk '{printf(" %s.pdb , %8.2f , %d \n", $28,$2,$27)}' >> ${OUTPUT}_summary.dat
````

### **`alphafold` : Deep learning in progress**   

The [AlphaFold](https://alphafoldserver.com/) tool, which revolutionized the field of protein structure prediction by applying deep learning (DL) methods, has been  expanded to include nucleic acids, offering thus new possibilities for the *direct* $\ce{1D} \rightarrow \ce{3D}$ modeling of aptamer sequences. However, as shown in a recent [benchmark study](https://doi.org/10.1021/acssynbio.5c00196) on the modeling of aptamers with AlphaFold, the data sets  used to train the DL  models are heavily biased toward canonical, biologically evolved nucleic acid motifs. As a consequence, the authors warn also about the need for *further refinement to reliably predict complex, underrepresented structures* of aptamer molecules. This would be even more relevant for DNA aptamers due to the  scarcity  of ssDNA structures in the PDB. 

Interestingly, the statistical-thermodynamic algorithms adopted in the $\ce{1D} \rightarrow  \ce{2D} \rightarrow \ce{3D}$ pathway for aptamer modeling may help mitigate the bias present in the experimental structures. The reason is that the generation and scoring of 2D structures relies on simple but *general* physically-based assumptions.  Nonetheless, it is very likely that the DL-based methods will be much improved in the short run, either by adding more diverse structures to the training datasets and/or by using hybrid methodologies combining DL and physical methods.

## **The problem of guanine quadruplets  (G4s)**

In general, the thermodynamic 2D methods are primarily focused to predict canonical Watson−Crick base-pairing interactions with allowances for occasional G/T wobbles. Hence, they have difficulties and  usually fail  to predict complex motifs such as pseudo-knots and, especially,  G-quadruplexes (**G4s**) that arise from non-canonical contacts.  

The **G4s** are complex architectures built from typically two to four  **G-tetrads** stacked on top of each other and stabilized by $\pi \cdot \cdot \cdot \pi$ interactions.  Each G-tetrad is constituted by a squared planar arrangement of four G nucleobases. A **G-tract** is then a pile of G bases stacked  across the G-tetrads. As a result, G4s exhibit a four-stranded architecture that is normally very stable. The structural complexity of G4s is even more evident when considering the possible topology of the **connecting loops**  between the G-tracts (parallel, antiparallel and hybrid) and their various conformations (propeller, lateral, diagonal).    

[x]:![](PNG/G4.png)
<img src="./PNG/G4.png" width="720" height="320" style="display: block; margin: 0 auto">

**Fig 9** Left: ribbon model the 1XAV structure showing a stable G4 motif. Right: possible topologies and conformations of connecting loops.     

The G4s motifs are relatively common in genomic DNA what has allowed the development of bioinformatic methods aimed to predict the propensity of a given 1D sequence to fold into a G4. Therefore, if a sequence composition (or experimental evidence) suggests the presence of  G4 motifs, it would be convenient to predict the number and identity of the G4s by using the following web servers.  

* [pqsfinder](https://pqsfinder.fi.muni.cz/)
* [QGRS Mapper](https://bioinformatics.ramapo.edu/QGRS/index.php)
* [G4Hunter](https://bioinformatics.cruk.cam.ac.uk/G4Hunter/)

These tools  give similar but usually not identical answers,  what represents a challenge in molecular modeling by augmenting the number of potential models to be investigated. Moreover, it turns out that [3D-NuS](https://project.iith.ac.in/3d-nus-qplex/index.html) is currently the only tool capable of generating initial 3D structures for G4s. Several options must be passed to this tool to specify the number and identity of G-tetrads,  the topology of connecting loop and the sequence of 3 loops connecting the terminal quartets. Note that additional model building operations would be needed to obtain 3D geometries of the rest of the sequence and that depending on the previous information this approach can lead to a high number of initial 3D models!

## **Improvement of the initial 3D models: `APTAMD` protocol**

###  **Initial 3D models: sampling is needed**   
Assuming that an aptamer sequence leads to a not-too-complicated fold, the reliability of the *initial* 3D models will be small, regardless of the computational details of the  $\mathrm{1D}  \rightarrow \mathrm{2D} \rightarrow \mathrm{3D} $ protocol (or direct $\mathrm{1D} \rightarrow \mathrm{3D} $ conversion). Several reasons explain this situation: 

* As above mentioned, although computationally efficient, the thermodynamic algorithms can miss important non-canonical 2D motifs.
* The initial 3D models may be  more or less biased towards genomic-like DNA segments that are most abundant in structural databases. This bias would be more pronounced in the case of DL-models for direct $\mathrm{1D} \rightarrow \mathrm{3D} $.
* Aptamer sequences may be compatible with different 2D forms leading to structurally distinct 3D models.
* Methods to build 3D RNA structures necessarily employ approximated scoring functions  and usually present structural defects and steric clashes. *Explicit* effects by solvent molecules and counterions are not considered.
* Virtual mutation of RNA structures to DNA does not relax the  backbone conformation and  may result in   misaligned base pairing and base stacking contacts.   
* In any case, **aptamer ssDNA molecules are intrinsically flexible systems** and, therefore, molecular modeling tools should be also employed to study their conformational properties.
  
Eventually, most of the limitations of an initial 3D model derived from `RNAComposer` or other builders can be related  to its *static* character. To overcome this problem using molecular modeling tools, the simplest choice is to use the **Molecular Dynamics** (MD) or MonteCarlo (MC) methods to generate realistic *ensembles* of biomolecules in aqueous solution. Some authors consider that the MD or MC methods add one more *dimension* to the modeling pathway (i.e.,  $\mathrm{3D} \rightarrow \mathrm{4D} $), what might be reasonable since  MD methods  do evolve 3D structures throughout a certain  simulation time. Nonetheless, it seems more adequate to employ terms like *MD-based 3D models* or just **MD models**.       

### **MD workflows for aptamer model building: a steep learning curve** 

The benefits of using MD methods in  aptamer model building were clearly shown in a 2017 paper by [Jeddi and Sainz](https://doi.org/10.1038/s41598-017-01348-5), in which the *de novo* predictions of DNA hairpin structures were systematically improved thanks to MD simulations.  As a matter of fact, during the last 15 years, there has been an increasing number of studies reporting the *in silico* characterization of DNA aptamers using MD simulations. 

[x]:![](PNG/pubs.png)
<img src="./PNG/pubs.png" width="640" height="320" style="display: block; margin: 0 auto">

**Fig 10** Number of publications reporting MD results on aptamer systems.

Unfortunately, chemists used to employ bioinformatic tools at web servers face substantial problems to start using MD methodologies. Among such difficulties, we have:

* There are no hardware concerns when using web servers.  For MD simulations, it is the opposite!
* Web servers require zero installation, but MD software is notoriously complex to install and manage.
* Web tools handle a few input options. Preparation of MD input is (at best) semiautomatic and conceptually demanding.
* Web tools return final results in the form of tables or figures. Post-processsing and  analysis of MD simulations requires customization and script writing.
* Bioinformatics methods give static answers, but interpreting MD ensembles requires statistical thinking and understanding convergence and sampling measurements.
* Users have limited control on the quality of web server outputs. In contrast, they must assume full responsibility of  MD settings and MD data.
* To avoid black-box usage, MD users need some understanding of statistical mechanics and physics.    

### **APTAMD: MD driver scripts and MD protocol**

Some of the obstacles that lie ahead on the application of MD methods may be overcome with the help of **driver scripts** that implement well-defined and reproducible  workflows for MD preparation, production and/or analysis.  In this way, a driver script encapsulates previous computational experience, acting thus as a bridge between the simplicity of web tools and the actual complexity of MD methodologies. 

**APTAMD** is a particular collection of MD driver scripts that has been developed *specifically* for simulating DNA aptamer biomolecules. It is also a **computational protocol** for model building defined by a sequence of simulation stages driven by the APTAMD `do_....` scripts (excepting for the first stage) as well as by the parameters/settings  encoded within the scripts:

1.  Construction of initial 3D models by the $\mathrm{1D}  \rightarrow \mathrm{2D} \rightarrow \mathrm{3D} $ approach.
2. *Edition* of initial 3D models driven by `do_aptamer_edition`.
3. *Enhanced* MD simulation on the $\micro$s *timescale* driven by `do_runmd` followed by the selection of best candidate(s) for equilibrium sampling ( `do_struct` and `do_rwgamd`). 
4.  Equilibrium MD simulation on the $\micro$s *timescale* (`do_aptamer_edition` + `do_runmd`). 
5.  Batch of structural and energetic analysis to determine the convergence degree of the simulations, the most likely aptamer conformations and the relative stability of competing models (`do_struct`, `do_sconform` , `do_mmpbsa`).
