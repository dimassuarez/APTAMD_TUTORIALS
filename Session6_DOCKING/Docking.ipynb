## **Prediction of non-covalent complexes: Docking algorithms** 
###  Molecular Recognition
Aptamers bind to their targets with high affinity and specificity. Molecular recognition is a combination of shape complementarity (*lock-and-key*) with a variable degree of structural adjustment (*induced-fit*) to create specific interactions with the target that results in  stable non-covalent complex. The non-covalent  forces stabilizing this complex include:
* **Hydrogen Bonding** Key in forming specific base pairing or H-bonds with polar groups of the target.
* **Electrostatic Interactions** Charged phosphodiester backbones and bases in the aptamer often interact with positively charged regions on the target.
* **Hydrophobic Interactions & $\pi \cdot\cdot\cdot \pi$  Stacking** Hydrophobic interactions, particularly with base aromatic rings, may play a significant role in binding stabilization.
* **van der Waals Forces** Weak vdW interactions, which  operate at short range, contribute to the specificity when the aptamer closely fits the target.

### Molecular Docking
[Molecular docking](https://www.annualreviews.org/docserver/fulltext/biochem/93/1/annurev-biochem-030222-120000.pdf?expires=1779727551&id=id&accname=guest&checksum=3EAE1CD8B1D38021CE3248034827A83E) is a key computational method designed to *dock* a ligand to the structure of a receptor. Traditionally, it has been used to predict the interaction between macromolecular targets (i.e., proteins and nucleic acids) and small ligands (<100 atoms), but it can be also used to study protein-protein or protein-DNA complexes. Most molecular docking methods assume that the binding conformation corresponds to an energy minimum and that the shape complementarity between receptor and ligand is a key factor for binding (i.e., *lock-and-key*). But binding can also involve local structural rearrangements in the receptor and/or in the ligand (*induced-fit*).

The term **blind docking** refers to docking a ligand to the whole surface of a receptor without any prior knowledge of the target pocket. In this sense, docking target to DNA aptamer molecules is generally a case of blind docking.  

Docking methods generally have two components: 

1. **Sampling procedure** to generate different orientations/conformations (poses) for the ligand bound to the receptor,
2. **Scoring function** to *estimate* the binding energy $\Delta_{bind}G$ or affinity of the different poses. The pose with the most favorable score is the predicted complex, and the score of this pose is the associated $\Delta_{bind}G$.


#### 1- Sampling or pose generation

Different search algorithms can be used to position and orient the ligand with respect to the receptor by means of translations and rotations. In such a case, the phase space for  **rigid-docking** has only 6 degrees of freedom (DOF). However, finding the optimal *fit* often requires to include flexibility in the ligand and/or in the receptor. The shape of the ligand itself can be modified particularly due to internal rotations around single bonds. Optionally, local rearrangement of the receptor structure can be also incorporated to pose generation. But these *flexible-docking*  procedures significantly increase the computational cost of the docking calculation as the dimensionality of the phase space for docking is increased (i.e., 6 + $N_rot$).

Pose generation algorithms include:

* *Direct methods* that sample the search space at predefined intervals (i.e. deterministic methods like conformational search, fragmentation, and database search)
* *Stochastic algorithms*  like the Monte Carlo (MC) and Genetic Algorithm (GA) methods that make random changes to the ligand DOFs, using probabilistic criteria to accept or reject new poses.
* *MD simulations* targeted to explore receptor flexibility and ligand binding pathways. Typically, enhanced MD techniques are employed.

Among these  methodologies, the stochastic approaches are probably the fastest and most efficient methods to explore the docking phase space.

* **Q3** The next cell code  implements a GA algorithm designed to reproduce a given target text sequence from  an alphabet of symbols that consists of the following steps:
  1. A first generation consisting  in  population with  $N_{pop}$ individual text strings is generated by pure random selection of symbols.
  2. Loop over a given number of generations 
     1. Each individual text string is scored in terms of its similarity with the target. If target is matched then stop. 
     2. Only  top individuals survive for next generation (the *elite* individuals).
     3. A new population is generated from the elite individuals by means of genetic mechanisms of shuffling and mutation.
   * Run the cell code trying different values for the `TARGET`, `NPOP`, `MUT_RATE`... variables. Does the algorithm succeed in reproducing the target string?
   * Find the correspondence between the GA variables and parameters for the matching-text-segment problem and those for the molecular docking problem.      

````python
###  Example of how a GA algorithm works 
import random

TARGET = "APTAMD_IS_FINE"
ALPHABET  = "ABCDEFGHIJKLMNOPQRSTUVWXYZ_"
NPOP   = 10      # Population size (number of individuals)
NGEN   = 5000    # Number of generations 
N_ELITE  = 4     # Number of top individuals to be kept (elitism)
MUT_RATE = 0.10  # Mutation rate
pfreq  = 50      # Print frequency

# 1. Initialize population  Random choice of ALPHABET elements
population = []
for i in range(NPOP):
    individual = ""
    for j in range(len(TARGET)):
        individual = individual + random.choice(ALPHABET)
    population.append(individual)

# Main generation loop
for gen in range(1, NGEN + 1):

    # 2. Calculate scoring for each individual
    fitness_scores = []
    for ind in population:
        score = 0
        for i in range(len(TARGET)):
            if ind[i] == TARGET[i]:
                score = score + 1
        fitness_scores.append(score)

    # 3. Find the best individual and its sccoring fit
    best_ind = population[0]
    best_fit = fitness_scores[0]
    for i in range(1, NPOP):         
        if fitness_scores[i] > best_fit:
            best_fit = fitness_scores[i]
            best_ind = population[i]

    if ( int(gen / pfreq) * pfreq == gen ) or ( gen == 1 ) :  ## 
        print("Gen " + str(gen) + ": " + best_ind + " (Fitness: " + str(best_fit) + ")")

    if best_ind == TARGET:
        print("Gen " + str(gen) + ": " + best_ind + " (Fitness: " + str(best_fit) + ")")
        print("Target reached!")
        break

    # 4. Selection: pick the top N_ELITE parents with the highest scores
    # Elements being sorted are indices (0, 1, 2, …)--> range(NPOP) 
    # but they are sorted by the fitness score  key= ( lambda i : fitness_scores[i] )
    ranked = sorted( range(NPOP), key= ( lambda i : fitness_scores[i] ), reverse=True)
    parents = []
    for i in ranked[:N_ELITE] :
        parents.append(population[i])
    
    # 5. Breeding and mutation to fill the next generation
    # Keep the N_ELITE parents unchanged (elitism)
    next_gen = []
    for p in parents:
        next_gen.append(p)

    while len(next_gen) < NPOP:
        # Choose two parents randomly from the elite pool
        p1 = random.choice(parents)
        p2 = random.choice(parents)

        # Crossover: random split point                       
        split = random.randint(1, len(TARGET) - 1)
        child = ""
        for i in range(len(TARGET)):
            if i < split:
                child = child + p1[i]
            else:
                child = child + p2[i]

        # Mutation: 10% chance to mutate one random character
        if random.random() < MUT_RATE:
            mut_idx = random.randint(0, len(TARGET) - 1)     
            mutated_child = ""
            for i in range(len(TARGET)):
                if i == mut_idx:
                    mutated_child = mutated_child + random.choice(ALPHABET)
                else:
                    mutated_child = mutated_child + child[i]
            child = mutated_child

        next_gen.append(child)

    population = next_gen                        
````

#### 2- Scoring Function
The second ingredient of traditional docking procedures is the **scoring or affinity prediction** function. Scoring functions take as input the coordinates of a binding pose to evaluate the energetic feasibility of ligand binding to the receptor in that conformation. Thus, a scoring function is used *to rank* the poses to finally select a stable and energetically favorable binding mode(s). 

Scoring functions fall into [several categories](https://www.ijpsjournal.com/article/A+Comprehensive+Review+of+Molecular+Docking+Principles+Methods+Applications+and+Future+Directions#) based on their underlying mathematical and physical principles. 

* **Physics-based or force-field** functions  use classical mechanics (MM) force fields to calculate the sum of non-bonded interactions (electrostatics and van-der-Wals). 
* **Empirical** functions estimate binding energies by summing weighted individual energy terms (hydrogen bonds, hydrophobicity, rotor penalties) with coefficientes derived from experimental binding data.
* **Knowledge-base** functions include potentials  derived from statistical analyses of known 3D structures of receptor/ligand complexes. 
* **Machine-learning and deep-learning** models use non-linear algorithms (neural networks) trained on structural and affinity datasets.  The [alphahold server](https://alphafoldserver.com/)  can also model  structures consisting of DNA and ligand molecules.  

Additionally, scoring functions often include ligand-only terms to account for the entropic penalty of constraining the ligand in the binding site and the enthalpic penalty of the ligand adopting the shape specified by the candidate binding pose (i.e., ligand *strain*).

Computationally, a good scoring function must be *cheap*, not only to rank all the poses generated by the sampling procedure for one ligand, but also to screen through *huge* datasets of ligand molecules targeted against the same or similar receptors.

## **AutoDock Suite** 

The core functionality of the AutoDock Suite is developed by the following three programs:
* **AutoDockTools**,  a graphical user interface designed specifically for setting up AutoDock and AutoDock Vina calculations. It allows you to prepare coordinate files for the receptor and the ligand, define rotatable bonds, define the search space on the receptor (i.e., the *grid box*), create input parameter files for the docking programs, and analyze the results. We are going to use several **preparatory python scripts** included in these tools: `prepare_receptor4.py`, `prepare_ligand4.py`, `prepare_gpf4.py`, and `prepare_dpf4.py`.
  
* **AutoDock Vina**, a docking program based on a hybrid scoring function that combines empirical data with knowledge-based terms (i.e. distance and contact-based atom type interactions) to improve the accuracy of pose predictions, and a rapid gradient-optimization conformational search that is very fast (often 100x than AutoDock 4!) and effective for most ligand-protein systems.

* **AutoDock4**, includes two main programs and provides user-tunable options for systems with special challenges. It  combines a Lamarckian genetic algorithm search method with a physics-based scoring function. This function estimates the binding affinity by summing pairwise interaction energies: Lennard-Jones 6-12 potentials for van der Waals interactions, Coulombic electrostatics with distance-dependent dielectric constants, and *directional* 10-12 hydrogen bonding potentials. Additionally, an empirical desolvation term accounts for the entropic and enthalpic costs associated with removing ligand molecules from aqueous solution, a critical factor for highly-charged systems often overlooked in simpler scoring functions. The desolvation term is modeled in terms of atomic volumes ($V_i$) weighed by  solvation parameters ($S_i$).    
$$ V_{{\mathop{\rm int}} }^{AutoDock} = {w_{vdW}}\sum\limits_{i,j} {\left( {\frac{{{A_{ij}}}}{{r_{ij}^{12}}} - \frac{{{B_{ij}}}}{{r_{ij}^6}}} \right)}  + {w_{X-H_i \cdot\cdot\cdot Y_j}}\sum\limits_{i,j} {f\left( {{\theta _{X - {H_i}\cdot\cdot\cdot{Y_j}}}} \right)} \left( {\frac{{{C_{ij}}}}{{r_{ij}^{12}}} - \frac{{{D_{ij}}}}{{r_{ij}^{10}}}} \right) +  $$
$$ + {w_{elec}}\sum\limits_{i,j} {\frac{{{q_i}{q_j}}}{{\varepsilon \left( {{r_{ij}}} \right){r_{ij}}}}} \,\, + {w_{solv}}\sum\limits_{i,j}^{} {\left( {{S_i}{V_j} + {S_j}{V_i}} \right)\,{e^{ - r_{ij}^2/2{o^2}}}} $$
The AutoDock scoring function also incorporates another empirical term that represents  the conformational entropy loss of flexible ligands upon binding . The weights of the energy terms ($w_{vdW}, \: w_{elec}, ...$) were originally calibrated using the structures and binding energies experimentally obtained for 188 protein/ligand complexes. Nonetheless, further [validation](https://doi.org/10.1021/ci800063v)   has shown that the AutoDock scoring function provides also accurate results for the docking of small molecules to nucleic acid targets. 
Therefore, although AutoDock4 is computational more demanding than other docking programs, it may be more reliable  for systems where charge-charge interactions or salt bridges are critical, such as DNA/RNA aptamers . Moreover, computational experience also suggests that  various atom types and charges that are parametrized within Autodock4  provide a balanced description of all interactions among different type of biomolecules.

## AutoDock4
It consists of two main programs that must be run consecutively:
* **AutoGrid** pre-calculates the energy environment of the receptor to speed up the docking process. The receptor is embedded in a three-dimensional Grid Box and the energy of interaction with each atom type in the ligand is calculated and assigned to each grid point, creating different energy maps: `receptor.C.map`, `receptor.O.map`, `receptor.N.map`, `receptor.d.map`, `receptor.e.map`, etc.

* **AutoDock** performs the docking of the ligand within the Grid Box that represent its interaction with the receptor. Ligand flexibility associated to rotations around single bonds can be added to rigid-body translations and rotations during pose generation.

Docking proceeds through the following steps: 
* **Step1: Prepare coordinate files** Two PDBQT files, one for the receptor and one for the ligand, should be generated. These are equivalent to the ones used by AutoDock Vina.
  
* **Step2: Energy maps with AutoGrid**  The input file for AutoGrid (extension `.gpf`, grid parameter file) can be prepared using a python script called `prepare_gpf4.py`. Required input include the name of the PDBQT files for  the receptor and for the ligand, the position of the grid center, its dimension and the spacing between grid points. When the binding site is unknown (blind docking), the grid dimension should be large enough to encompass the entire receptor and its ligand. However, the computational cost of computing the energy maps scales as $\mathcal{O}\left( N_{grid}^3 \right)$ and, in addition, a large grid box also increases nearly-exponentially the size of the phase space to be sampled. Therefore, the grid size determines the actual conformational sampling that can be achieved within reasonable computational time limits.
```bash
  prepare_gpf4.py -l ligand.pdbqt -r receptor.pdbqt -o receptor_ligand.gpf -p gridcenter=0.,0.,0. -p npts=172,254,156 -p spacing=0.333
```

*  **Step3: Docking using AutoDock.** The input file for AutoDock (extension .dpf, docking parameter file) can be prepared using a python script called `prepare_dpf4.py`. The docking parameter file lists the required grid maps previously calculated for each atom type in the ligand, the PDBQT file for the ligand, and other parameters that depends on the selected conformational search algorithm (GA=Genetic Algorithm; LGA=Lamarckian Genetic Algorithm; SA=Simulated Annealing; LS=Local Search). User can usually also select the number of docking runs (ga_run=50).
```bash
prepare_dpf4.py -l ligand.pdbqt -r receptor.pdbqt -o receptor_ligand.dpf -p ga_run=50
```
AutoDock writes the final docked coordinates into a `.dlg` file (docking log file).
  
*  **Step4: Evaluating the results of a docking calculation.** A clustering analysis is usually done on the configurations generated by  a docking calculation to identify similar poses. The AutoDock program incoporates an `analysis`option within the docking parameter file (`.dpf`) that activates cluster analysis, organizing docked conformations by their RMSD similarity and identifying the lowest-energy conformer within each cluster. This process helps identify consistent binding modes and, in many cases, helps locate near-native conformations, as highly clustered results are indicative of an exhaustive sampling of conformational space.



## **AutoDock4 within the APTAMD protocol** 

Since aptamers are quite flexible molecules docking calculations  should consider somehow their conformational variability. In the so-called [ensemble docking](https://blog.litefold.ai/ensemble-docking-vs-static-docking-when-does-protein-flexibility-matter/) approach, a ligand is docked against a collection of different conformations of the same receptor, rather than to a single rigid structure. Although each of the selected geometries for the receptor may remain fixed during the docking calculations, the simultaneous ranking of the docking poses for all the receptor configurations introduces in a effective manner a certain conformational sampling.   

Clearly,  the MD ensembles of aptamer molecules produced by the APTAMD protocol could be used to carry out ensemble docking.  A suitable set of MD frames would include *representatives* of the most-populated clusters,  but also of the scarcely populated ones to introduce a wider conformational diversity. To use the Autodock programs ,  it is necessary to prepare the coordiante and configuration files (`.pdbqt`, `.gpf`, `.dpf`), calculate grid maps, perform docking search and scoring, collect and rank the docking poses,  etc. To help automate these tasks, we have prepared two APTAMD scripts, `do_adck_templates.sh` and `do_adck.sh` that actually define a *docking protocol*.  

### The  APTAMD protocol for ensemble docking             

Starting from a given number $N_{rec}$ of MD frames (`.pdb` files) for the aptamer and $N_{lig}$ pdb files for the ligand molecule, the docking protocol consists of the following steps:

* Automatic preparation of the `PDBQT` and configuration files `gpf` and `.dpf` for grid map and docking calculations, respectively. AMBER topology (`.top`) and coordinates (`.crd`) files are generated too unless they are provided as input.  
* Distributed processing of the grid map calculations  over the $N_{rec}$ grid aptamer geometries. 
* Distributed processing of the `autodock4` calculations to yield $N_{pose}$ per MD frame (a total of $N_{rec}\times N_{lig} \times N_{pose}$ complexes).
* Distributed processing of structural relaxation of all the poses using the `sander` program.
* Finally, rescoring of the relaxed poses with the `autodock4` program. Only single-point calculations are now carried out on the relaxed geometries.

Sturctural  relaxation is quite convenient because grid-evaluated scoring functions  often fail to completely eliminate unphysical, high-energy contacts. These steric collapses can be fixed by means of an energy minimization followed by a new rescoring of the aptamer-ligand interaction in the minimized complexes. 

The  `do_adck_templates.sh` and `do_adck.sh`  scripts  organize, run and control input/output for all the  subtasks involved in the docking protocol.
#### **1- `do_adck_templates.sh`**

This  script  generates the input files required by AutoGrid and AutoDock. Input parameters include the name of two text files with the list of PDB files for the receptor and for the ligand and, optionally, the name of amber topology files for the receptor and for the ligand. An example of the input file (`input_adck.src`):
```
RECEPTOR_PDB_LIST="R.txt"  # Input file containing the list of PDB
                           # files with coordinates of the receptor
                           # No default value
LIGAND_PDB_LIST="L.txt"    # Idem for Ligand
LIGFORMAT="FLEX"           # Activate torsional degrees of freedom

NPROCS=8                  # Number of cores to be used. Def all available.

GRID="0.333"               # Grid size of the Autogrid maps. Def =0.333 Ang

GA_RUN="50"                # Number of GA runs to request in the ADCK calcs.
                           # NOTE: A total of NREC x NLIG x GA poses will be
                           # produced. Def = 50


RELAX="YES"                # Prior to ADCK atom typing, relax the geometries
                           # of the receptor/ligand reference geometries.
                           # YES is recommended


TOPOLOGY_REC="apt.top"     # Optional Amber topology file for receptor
TOPOLOGY_LIG="AMP.top"     # Optional Amber topology file for ligand
```

The script can be submitted with `qapta do_adck_templates input.src`. It performs the following tasks:

* Selects an optimal grid size based on the dimensions of the largest conformation of the receptor and ligand, thus ensuring an adequate search space for all the conformations.
* Relaxes the geometry of the receptor and of the ligand for a proper assigment of atom types and charges during the generation of PDBQT files.
* Generates a PDBQT file for the receptor using the `prepare_receptor4.py` script. Guanine charges must be corrected in the resulting PDBQT file to obtain a global -1 value.
* Generates a PDBQT file for the ligand using the `prepare_ligand4.py` script. Ligand torsions can be active/inactive in the resulting PDB file depending on the selected option.
* Generates the Grid Parameter File (.gpf) that serves as input for AutoGrid with the `prepare_gpf4.py`script.
* Generates the Docking Parameter File (.dpf) that serves as input for AutoDock with the `prepare_dpf4.py` script.


#### **2- `do_adck.sh`**. 

The second script drives the time consuming calculations which include AutoGrid, AutoDock and AMBER MM energy minimizations. 

* Creates a directory for each conformation of the receptor in the RECEPTOR_PDB_LIST file (REC_1, REC_2, REC_3, ...), and within these, different subdirectories for the different conformations of the ligand in the LIGAND_PDB_LIST file (LIG_1, LIG_2, LIG_3,..).
* Runs AutoGrid for each conformation of the receptor, generating the required energy maps in the selected Grid Box.
* Runs AutoDock for each conformation of the receptor and of the ligand generating a docking`.log` file (`receptor_ligand.dlg`) that contains the required poses. Thus each `do_adck` run generates `NREC x NLIG x GA_RUN` conformations for the recpetor/ligand complex.
* Combines the coordinates of the receptor and of the ligand in the docking log file to generate complexes that are edited and minimized with `sander`.
* Rescores the relaxed complexes with the AutoDock4 scoring function and provides list (`SCORING.dat`) where the most favored poses are ranked according to their intermolecular interaction energies predicted by AutoDock.   

## **Parameterization of small (ligand) molecules** 

Molecular docking calculations using the `APTAMD protocol` employ Amber topology files for both the receptor and the ligand, in order to perform energy minimizations before atom typing (`do_adck_templates`) and to eliminate steric collisions in the resulting docking complexes (`do_adck`). Due to the extraordinary diversity of small molecules that are targets of aptamers, ligand parameterization can be a preliminary step to molecular docking calculations. Although an automated parameterization script has been developed and user in our group, it needs to be largely rewritten to use more up-to-date software tools and enhance the parameterization protocol. Therefore, we provide here a simple recipe using the tools available in the AMBER package.

The `antechamber` program is the main tool for ligand parameterization. It performs different actions: 

1. Recognizes atoms and bond, and assigns atom types accordingly using the GAFF2 (General Amber Force Field)
2. Computes atomic point charges using  QM data. In general, charges are obtained from a restrained fit to the QM electrostatic potential.  antechamber can accept QM data produced by a previous QM calculation (most recommended option) or can execute on-the-fly a semiempirical QM calculation (good for docking purposes).
3. Generates initial parametrization files .

After running `antechamber`,  a second program, `parmchk` must be employed to detect missing force field parameters and supply reasonable and similar substitutes.

```bash

```
`antechamber` creates many intermediate files, `ANTECHAMBER_*`, that can be safely deleted.  Charges and atom type assignments are specified in the `.mol2` file. 

3. Next run `parmchk` to complete the parameterization:
   
```bash
 parmchk2 -i AFB1.mol2 -f mol2 -o AFB1.frcmod
```
the new `AFB1.frcmod` file contains a list of additional BAD parameters (bond, angle and dihedral). 

4- Create a `edit_AFB1.src` file with the following lines:
```bash
# Force Field data
source leaprc.gaff2
loadAmberParams AFB1.frcmod
AFB=loadMol2 AFB1.mol2
# .top and .crd generation
mol=loadpdb AFB1.pdb
check mol
saveamberparm mol  AFB1.top AFB1.crd
quit
```

4- Use tLEaP to generate the topology (`.top`) and coordinates (`.crd`) files for the AFB1 molecule:

```bash
tleap -f edit_AFB1.src
```


## **Limitations of the docking approaches** 
Molecular docking calculations are fast computational methods that provide a list of different receptor/ligand complexes ranked by energy. However, several key factors limit the accuracy of the predicted docking scores, including:

* **Rigid receptors** The receptor is treated as rigid or semi-rigid (e.g., side-chain flexibility) in docking calculations, but ligand binding usually triggers conformational changes in the receptor (induced fit). A static apo (unbound) pocket may not be pre-formed in the exact active conformation, making bonding in the correct location difficult (i.e. wrong binding site). Tiny changes in residue positions in the binding site usually lead to drastically different binding orientations.
  
* **Aproximate scoring functions** Scoring functions should be able to distinct the true binding modes from all the other alternative modes. But docking scores are mathematical approximations rather than accurate and rigorous free energy calculations. So frequently, there is poor correlation between the top numerical scores and biological assay data (e.g., $R^2\sim 0.6-0.7$ at best). In addition, the best-scoring pose does not always correspond exactly to the real conformation.

* **Environment effects** Water is usually treated implicitly in docking calculations, but water molecules often act as bridges, forming hydrogen bonds between the receptor and the ligand. Metal ions such as $\ce{Na+}$, $\ce{K+}$, and $\ce{Mg^{2+}}$  are usually not included in docking calculations. However, they are crucial for the folding of aptamers into their active 3D tertiary structures, also affecting to the strength of the bond to its target.

*  **Entropic effects** Docking often fails to accurately account for the entropy lost when a flexible molecule is frozen into a specific binding pose, or the entropy gain when solvent is displaced from binding sites. 

As a result, **top-scoring candidates selected from docking calculations should be validated**. Eventually, such validation may be achieved through  long-time MD  simulations to check whether the ligand remains attached or detaches from its receptor in the selected binding poses.....

### Selected aptamer
As a case study, we selected a DNA aptamer that binds AMP (Adenosin MonoPhosphate). The [NMR solution structure](https://doi.org/10.1016/s1074-5521(97)90115-0) of the aptamer···AMP complex (PDB Code: 1AW4) shows two AMP molecules intercalated at adjacent sites within a rectangular widened minor groove. In these sites, two recognition G9···AMP1 and G22···AMP2 mismatches are flanked by sheared G···A and reversed Hoogsteen G···G mismatch pairs. According to experiments, **the aptamer undergoes a conformational change upon binding**, *zippering* up to form a stable complex with well-defined binding sites.

### Docking with AutoDock Vina
Let's start using AutoDock Vina, a fast and reliable program for predicting protein/ligand complexes.

1- Create a directory named `VINA` within your `DOCKING` directory. Edit the 1AW4 pdb file and save the atomic coordinates of the first NMR model (`MODEL 1`) in a separate file (e.g., `1AW4_1.pdb` located in the `VINA` directory). Move to the `VINA` directory and edit the new file to write  the coordinates of the aptamer receptor (residues 1-27) and of the first AMP ligand molecule (residue 28) in the  `receptor.pdb` and `ligand.pdb` files, respectively.   

2- Use the python scripts provided by AutoDockTools to generate the `receptor.pdbqt` and `ligand.pdbqt` files (these tools are available at the command line):
```
     prepare_receptor4  -r receptor.pdb
     
     prepare_ligand4  -l ligand.pdb

```
3- Inspect the `ligand.pdbqt` file  to check that, by default, rotation around single bonds will be allowed during the docking calculation. 

4- Create an input file for AutoDock Vina (`input_vina.txt`) including the required information:
```
   receptor = receptor.pdbqt
   ligand = ligand.pdbqt

   center_x = valueX
   center_y = valueY
   center_z = valueZ

   size_x = value_sX
   size_y = value_sY
   size_z = value_sZ

   exhaustiveness = 8
```
valueX, valueY, and valueZ should be replaced by appropriate values to center the Grid Box close to binding site. In aptamers, the binding site is usually unknown (blind docking) so it is advisable to place the center of the grid at the center of mass of the receptor. These coordinates and the dimension of the receptor can be obtained using the center_geom program:  
```
      $APTAMD/AUXTOOLS/center_geom -noatm -i receptor.pdbqt -o receptor.cen
```
Check the ouput `receptor.cen` file to obtain the required information.

### Ensemble Docking with AutoDock4

AutoDock4 is computationally more expensive than AutoDock Vina, but its AMBER-like scoring function seems more appropriate for docking calculations in aptamers due to the explicit consideration of electrostatic interactions. In addition, thanks to previous MD simulations performed for the solvated aptamer and for the AMP ligand, we can introduce flexibility both in the receptor and in the ligand using an ensemble docking approach. To perform the calculations, follow these steps:

1- Create new directories:  `DOCKING/AutoDock` and `DOCKING/AutoDock/PDBs`.

2- Move to `DOCKING/AutoDock/PDBs` and copy all the `aptamer*` and `AMP*` files stored in `/opt/data/apta`. These files correspond to cluster representatives obtained from an MD simulation of the aptamer and of the AMP molecules in water and the corresponding topology files.

3- Move to the `DOCKING/AutoDock` directory. Before using  the first script `do_adck_templates.sh` to generate the information and the input files for AutoGrid and AutoDock, we need to create several list files:

* **`R.txt`** contains the list of PDB files of the receptor (aptamer) . It can be created with the following sequence of linux commands:
  ```
  ls PDBs/aptamer*.pdb >R.txt
  ``` 
* **`L.txt`** contains the list of PDF files to be used for the ligand (AMP) in the docking calculations. As in the previous point:
  ```
  ls PDBs/AMP*.pdb >L.txt
  ```
*  We also need the `aptamer.top` and `AMP.top`AMBER topology files for the receptor and  ligand, respectively. You can copy/move them from `PDBs/`.
  
*  **`input_adck.src`**. This input file defines the important settings for the docking calculations. Enter `do_adck_templates` at the command-line to obtain a brief explanation and a list of the main options to run the script. Thus the file `input_adck.src` should include, at least, the definition of the following variables:

```
RECEPTOR_PDB_LIST="R.txt"  # Input file containing the list of PDB
                           # files with coordinates of the receptor
                           # No default value
LIGAND_PDB_LIST="L.txt"    # Idem for Ligand

NPROCS=8                  # Number of cores to be used. Def all available.

GRID="0.333"               # Grid spacing of the Autogrid maps . Def =0.333 Ang

GA_RUN="10"                # Number of GA runs to request in the ADCK calcs.
                           # NOTE: A total of NREC x NLIG x GA poses will be
                           # produced. Def = 50

RELAX="YES"                # Prior to ADCK atom typing, relax the geometries
                           # of the receptor/ligand reference geometries.
                           # YES is recommended

TOPOLOGY_REC="aptamer.top"     # Optional Amber topology file for receptor
TOPOLOGY_LIG="AMP.top"     # Optional Amber topology file for ligand
```

4- Submit the preparatory  task done by `do_adck_templates` to generate all the required files for the `AutoGrid` and `AutoDock` calculations.
 
```bash
do_adck_templates input_adck.src 
```
Have a look (`ls *`) to the files generated by `do_adck_templates`. The most important files are: `receptor.pdbqt` and `ligand.pdbqt` with the AutoDock atom types and Gaisteger atomic charges for both molecules,  the output files `receptor.out` and `ligand.out` for the Amber minimization carried out before  atom typing, and the topology `complex.top` file needed for relaxation of the docking poses. The APTAMD tools fix the `.pdbqt` files concerning atom/residue names and Gasteiger charges assigned to the DG residues. 

5- Submit the docking calculation to the queue:
 
```bash
qapta do_adck input_adck.src  #   qapta is the submission command in our local cluster 
```
This docking calculation will take about 10 minutes to complete.
