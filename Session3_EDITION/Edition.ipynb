### **What is a PES?** 
The PES is a consequence of the Born-Oppenheimer (BO) approximation that  involves a total separation of electronic  $\bf{r}$  and nuclear $\bf{R}$ degrees of freedom.  By adopting the BO approx. , one can first solve the *electronic* time-independent Schrodinger equation at any fixed $\bf{R}$ molecular geometry, 
$$ \hat{H}_e ( {\bf{r;R}} ) \Psi_{e,k} ( {\bf{r;R}} )  = E_{e,k} \Psi_{e,k} ( {\bf{r;R}} ) $$ 

where $\hat{H}_e ( {\bf{r;R}} )$ is the electronic Hamiltonian collecting all kinetic and potential energy terms involving electronic coordinates, while $E_{e,k} $ and $\Psi_{e,k} ( {\bf{r;R}} )$ are the electronic energies and wavefunctions for the electronic states $k=0,1,2, ...$. When the total wavefunction is  expressed  as $\Psi ( {\bf{r,R}} ) = \sum_k  \phi_k({\bf{R}}) \Psi_{e,k} ( {\bf{r;R}} )$, the BO approximation gives a new Schrodinger equation that depends on the coordinates of the nuclei,

$$  \left[ \hat{T}_N ( {\bf{R}} ) + \hat{V}_N ({\bf{R}}) + E_{e,k} ( {\bf{R}} ) \right] \phi_k ( {\bf{R}} )  = E \phi_k ( {\bf{R}} ) $$ 

In this equation,  the electronic energy $E_{e,k} ( {\bf{R}} )$ accounts for the *instantaneous* adjustment of electrons to changes in molecular geometry. Hence, $E_{e,k} ( {\bf{R}} )$ can be interpreted  as an  effective electronic potential acting on the nuclei and, therefore, the sum $\hat{V}_N ({\bf{R}}) + E_{e,k} ( {\bf{R}} )$ turns out to be the potential that ultimately determines nuclear motions or, in other words, the geometry and dynamics of molecular systems.  

**The sum of the Coulombian nuclear repulsion energy and electronic energy defines the PES required to describe nuclear motions either classically or quantum mechanically**.

 **PES**    $\:\:\: \longrightarrow   \:\:\:  E\left( \bf{R} \right ) =  \hat{V}_N ({\bf{R}}) + E_{e,k} ( {\bf{R}} )$ 

By construction, $  E\left( \bf{R} \right ) $ **is a continuous multivariate function of the nuclear coordinates** that, in turn, define the geometry of a molecule or of a set of molecules. In principle, $  E\left( \bf{R} \right ) $ can describe all situations of chemical interaction, whether they are the formation/breaking of covalent bonds or non-covalent attraction/repulsion. Fortunately, many chemical properties can be explained by studying the PES of the fundamental electronic state ($k=0$) so that PESs are characterized by default at the fundamental electronic state.   



## **Molecular Mechanics**

The term *Molecular Mechanics* (MM) refer to an intuitive model of molecular structure that consists of a set of rigid spheres joined together by specific forces, what may remind us of ball-and-stick 3D images. To perform a MM calculation, we  need a  **MM potential** (also known as *force field* ) to estimate the change in a PES $E\left( \bf{R} \right )$  as a consequence of small variations in bond distances and angles and, above all, of the conformational changes associated with internal rotations around chemical bonds and the formation/breaking of non-covalent interactions. Traditionally, the MM methods have adopted simple mathematical expressions, known *a priori*, to calculate both the potential energy $E\left( \bf{R} \right )$ and its gradient (i.e., the *force* acting on the atoms ${\bf{F}} =-\nabla E$). To be of practical utility, we will see that the MM energy formulas  include *parameters* that must be previously optimized from experimental and/or theoretical data. 

The **main advantages** of the MM methods are:
* MM calculations  are computationally very fast (expensive quantum mechanical calculations are avoided!).  
* MM methods allow us to run *classical* MD simulations on the $\mu$s time scale of systems with up to $\sim 10^6$ atoms and including long-range effects in an accurate manner!

Their obvious **major limitations** are:  
* MM methods do not provide any *electronic* property (the $e^{-}$ density is taken into account only in an implicit manner by  *atomic charges or multipoles*).   
* The scope of application of the MM methods is predetermined by the actual form of the MM potentials and the specific parameters being used.

### **MM potentials: combining chemical intuition and classical physics**

The most common assumptions considered in the development of MM potentials for biomolecular simulations are: 

1. Two types of energy (and force) contributions are distinguished :
$$ {E_{MM}} = E_{bonding} + E_{nonbonding} $$
    where $ E_{bonding}$ collects the energy associated with covalent bonds while $ E_{non-bonding}$ accounts for  non-covalent interactions. Both energy terms should be explicit functions of  $\bf{R}$ (i.e. molecular coordinates).  

2. $ E_{bonding}$ takes into account just the effects of small fluctuations in  molecular geometry due to stretching, bending and torsional  motions.  If we  neglect  cross-terms  between the  internal motions, we have:  
$$E_{bonding} = \sum\limits_{bonds} E_{stretching}  + \sum\limits_{angles} E_{bending}  + \sum\limits_{dihedrals}  E_{torsion}  $$
    Note that, prior to the calculation of $ E_{bonding}$, it is necessary to specify the molecular connectivity (i.e. the molecular *topology*) and that $E_{bonding}$  just measures the energy required to distort a molecule in a particular way with respect to its reference geometry.
    
3. The non-covalent energy, $E_{nonbonding}$, is calculated as a sum of atomic pair contributions, distinguishing between van der Waals (vdW) and electrostatic forces.   
$$ E_{nonbonding} = \sum\limits_{atom \: pairs} \left( E_{vdW} + E_{elect} \right) $$
    A  criterion is needed to select which pairs of atoms are involved in nonbonding interactions. Most commonly, $ E_{nonbonding}$ is calculated  between atoms belonging to different molecules and between atoms of the same molecule *separated*  by at least three covalent bonds (i.e. a topological distance $\geq 1-4$ ).

### **Description of the AMBER potential**

The [AMBER](https://ambermd.org/AmberModels.php) (*Assisted Model Building and Energy Refinement*) potential  was designed to simulate conformational energies and molecular interactions in proteins and nucleic acids, but it can be applied to any organic molecule bearing   functional groups similar to those in biomolecules. The  AMBER functional form is particularly simple:

$$ E_{AMBER} = \sum\limits_{bonds} k_r \left( r - r_{eq} \right)^2  + \sum\limits_{angles} k_\theta \left( \theta  - \theta _{eq}\right)^2   \:\: + \sum\limits_{dihedrals} \sum\limits_{n = 0}^m  k_n   \left[ 1 + \cos \left( n \phi  \right) \right] \:\: + $$
$$ + \sum\limits_{i < j} \left[ \:\: \left[  4\varepsilon _{ij}  \left( \frac{\sigma_{i,j}}{R_{i,j}} \right)^{12} - \left( \frac{\sigma_{i,j}}{R_{i,j}} \right)^{6}  \right] \:\: + \frac{1}{ 4\pi \varepsilon } \frac{q_i q_j }{R_{ij}} \right] $$

Similar expressions are also employed by other biomolecular force fields like CHARMM and OPLS. 

For proper usage of AMBER force fields, it is important to recognize the following  characteristics  of $E_{AMBER}$ :   

1. **Stretching and bending** motions are described by  a **harmonic potential**,  a physically-correct model around equilibrium geometry. Each harmonic term contains  two parameters: the force constants ( $k_r$ and $k_\theta$ )  and  the reference values (distance $R_{eq}$  and  angle  $\theta_{eq}$).

<img src="./PNG/harmonic.png" width="720" height="540" style="display: block; margin: 0 auto">

**Fig 2**  Stretching and bending potentials between C($sp^3$) atoms.
   

2. **Dihedral (torsional) angles** $\phi$ defined by four atoms (e.g. X-C-C-X) are the natural coordinates to describe the rotational motions around the central  bond. Given that $E_{torsion}$ must be a periodic function of  $\phi$, AMBER represents $E_{torsion}$ as a Fourier expansion of cosines. The cosine terms include parameters $k_{n}$ that determine the barrier to rotation around the central link.
   
[x]:![](PNG/torsion.png)
<img src="./PNG/torsion.png" width="720" height="540" style="display: block; margin: 0 auto">

**Fig 3**  Torsion potential for internal rotation about  a single $\ce{C-C}$ bond.   
   Due to the critical role of  $E_{torsion}$, modern force fields incorporate also torsional energy *correction maps* (CMAP) that depend on two or more torsional angles and  are derived from QM data. For example, the CMAP for backbone motions of an amino acid is a set of tabulated  energy values on a regular grid of ($\phi$, $\psi$) angles.  Cubic spline interpolation is used to compute CMAP forces and energies at arbitrary values of $\phi$ and $\psi$ . Note, however, that CMAPs have not been yet developed for DNA systems.  

3. The **vdW energy** due to intermolecular or intramolecular ($\geq 1-4$)  interactions is represented by the computationally-efficient 6-12 form of the [Lennard-Jones potential](https://doi.org/10.1021/acs.jctc.4c00135). This pairwise potential plays a central role in molecular simulation. For a given pair of atoms, the LJ potential is defined by two parameters related to the position ($\sigma=2^{1/6}R_{min}$) and  depth ($\varepsilon _{ij}$) of the vdW energy minimum. Note that the LJ comprises both repulsive and attractive interactions that operate at  very-short-range ($R<\sigma$) and **short-range** ($R<9-10\mathring{A}$) distances, respectively. 

4. The **electrostatic energy** arises as a result of the distribution of the electronic and nuclear charge of molecules.  In AMBER, the electrostatic interaction energy is modeled as the interaction between ***fixed*  point charges** according to Coulomb's law (since the charges are fixed, this is a **non-polarizable** potential!).  Point charges $q_i$ usually centered on the atomic positions are treated as parameters of the force field. Compared with vdW forces, electrostatics gives rise to more intense interactions that operate both at short- and long-range distances and, for this reason, they are usually termed as **long-range** forces.

### **Developing AMBER parameters: an ongoing effort**

Since 1981 *dozens* of AMBER force field versions have been developed, evolving into specialized branches for proteins, nucleic acids, and small molecules that are under active development and testing. Such *evolution* is  fueled by:  
* The continuous improvements in computational algorithms and in hardware technologies,  expanding the capabilities of parameterization and validation of force fields. 
* The continuous enlargement of the reference datasets, comprising both experimental structures and high level QM data.  

In general, the  parameterization of force fields avoids a brute force strategy in favor of a workflow of sequential steps guided by chemical intuition. For AMBER, the parameterization stages are:

1. **Defining atom types**. To keep the total number of parameters within reasonable bounds,  we rely on the chemical concept of *functional group*  to define  *transferable* atom types, which not only refer to the atomic numbers, but also to their chemical environment. Thus, AMBER force fields distinguish up to 40 atom types distributed in a handful of elements (CNOPSH). 

1.	For each pair of atom types,  bond and angle parameters ($k_r,\: R_{eq}, \:...$) are taken as fitting parameters adjusted to reproduce equilibrium geometries and force constants taken from experimental and/or QM reference data. 

2.	The **Lennard-Jones (LJ) parameters** of each type of atom are optimized to reproduce the thermodynamic properties of simple organic substances. For example, the LJ parameters for the $\ce{C(sp^3)}$ and $\ce{C(sp^2)}$ atom types have been obtained by computer simulations of liquids ($\ce{CH4}$, $\ce{C2H6}$, $\ce{C6H6}$, ...), adjusting $R_{min}$ and $\varepsilon_{ii}$  to reproduce the densities and vaporization enthalpies $\Delta_{vap} H^\circ$. As vdW forces are non-specific interactions, it turns out that the LJ parameters are highly transferable among different molecules.

3. Atomic charges $q_i$ are obtained by the **Restrained Electrostatic Potential (RESP) method**, which fits atom-centered point charges to match  the molecular electrostatic potential (MEP) calculated over a large 3D grid mesh (see details [here](https://docs.openforcefield.org/projects/recharge/en/stable/users/resp.html)). The reference MEP data is obtained from QM calculations of the molecule or monomer of interest so that, in contrast with LJ parameters, the RESP charges are *specific* parameters. For macromolecules (proteins, nucleic acids, lipids and glycans), the various branches of the AMBER force field provide specific charges for each residue.

4. The final stage in the parameterization process optimizes the torsion parameters that complement the atomic charges and LJ parameters in order to determine the potential energy profile of internal rotations. The reference data are again obtained from  QM calculations that nowadays can predict very accurate rotational profiles and conformational energies for medium-size biomolecules.  This is a complex and critical parameterization step that still receives much attention by developers, leading to better and better  parameters for simulating the  backbone and side chain  conformational motions of peptides and nucleic acids.

### **AMBER force fields for DNA: Recommendations** 

The ff99 version of the AMBER force field for nucleic acids provides most of the parameters that are currently used in DNA simulations, including atom types, bond-angle parameters, the LJ parameters and RESP charges. However,  major improvements in the quality of the ff99-based force fields have been achieved by reparameterizing the torsional parameters for selected dihedrals (e.g. sugar ring puckering dihedrals,  $\beta$, $\chi$ and $\epsilon$ backbone torsions). According to various computational tests, the latest modifications termed as `bsc1`, `tumuc1` and `OL24` perform similarly and are valuable improvements over the previous versions (see the [AMBER manual](https://ambermd.org/doc12/Amber26.pdf) for further details). Nonetheless, it must be also noticed  that DNA force fields are tested with double-stranded DNA systems, including  *difficult* systems like A-tracts, CG-rich duplexes, Z-DNA and G-quadruplexes. Unfortunately,  specific assessments for aptamer simulations are scarce.  

* **Basing on our previous computational experience, the results of other validation studies and computational efficiency considerations,  we currently recommend the AMBER [`ff99+bsc1` force field](https://pmc.ncbi.nlm.nih.gov/articles/PMC4700514/)  to carry out MD simulations of ssDNA aptamers in combination with the TIP3P water model** (see below).

Of course, preferred force field choices may change in response to new validation and developments or depending on the actual system being simulated. For example, the following parameter recipe, ff19SB(protein)/ff99+OL25(DNA)/OPC(Water), may be more adequate  for complexes of aptamers with intrinsically-disordered peptides.
   
#### On the modified nucleotides (XNA)
XNA are synthetic nucleic acids bearing sugar and/or backbone modifications that are becoming quite relevant for aptamer development. Since the AMBER force fields only include parameters for natural nucleic acids, a new tool has been recently designed, [`modXNA`](https://modxna.chpc.utah.edu/) , that creates  Amber compatible parameterizations of XNA residues.  The last release of the  AMBER package contains  this tool, but currently the `APTAMD` driver script for aptamer edition handles only natural DNA residues.  



## **Solvation: MM representation of water solvent and counterions**

### **Simulating Liquid Water**  

A [*water model*](https://doi-org.uniovi.idm.oclc.org/10.1002/) consists of a MM potential and parameters aimed to reproduce the properties of *liquid water* (density, heat of vaporization, compressibility, self-diffusion coefficient, heat capacity, etc.). Of course, a realistic description of intermolecular interactions involving water molecules is required to describe hydration shell structure, but emphasis is placed on mimicking the bulk properties of water that largely influence the outcome of biomolecular simulations in aqueous solution. Although *continuum* solvent models (also known as *implicit* models)  have been proposed and are also utilized in molecular modeling, we discuss here only the case of ***explicit* water models** because they are essential to produce a reliable conformational sampling of flexible biomolecules.  

In explicit-solvent simulations,  waters account for over 90% of atoms  and, consequently, water models need to be fast and accurate. In this respect, the best alternative is to use  *rigid*  point-charge  models, which are widely used in biomolecular simulations. These models are defined by a set of vdW parameters ($A$ and $B$, assigned only to the O atom), fixed point-charges ($q_+$ and $q_-$, either located at or *shifted* from atomic positions) and two distances $R(\ce{O-H})$ and $R(\ce{H···H})$. The MM interaction energy between two waters is then expressed as:  

$$E \left( Wat1 \cdot\cdot\cdot Wat2 \right) = \sum\limits_{i}^\mathrm{on\:Wat1} \sum\limits_{j}^\mathrm{on\:Wat2} \frac{q_i q_j}{r_{ij}}  + \frac{A}{r_{OO}^{12}} - \frac{B}{r_{OO}^6}$$

Several models fit within this category, the most popular ones being  **TIP3P** and **OPC**: 
*  **`TIP3P`**  was derived by fixing the experimental geometry and using atom-centered point-charges. Despite of having been developed 40 years ago, `TIP3P` is still one of the most used  models because it is *cheap* and reasonably accurate.

*  **`OPC`**  is a four-point model in which the position of atoms and charges were freely optimized (preserving $C_{2V}$ symmetry). Since OPC is a 4-point model, it demands more computer resources than TIP3P. However, OPC reproduces very closely the reference water properties and there is growing evidence that OPC improves systematically the structural and energetic results of  MD simulations for all kind of biomolecular systems. Interestingly, a 3-point version (**OPC3**) is also available although it has been less tested. 
 
Although TIP3P is being overthrown by OPC, TIP3P in combination with ff99+bsc1 is still recommendable for nucleic acid simulations because: (a) TIP3P usage is supported by numerous computational studies; (b) [a former assessment](https://doi.org/10.1021/acs.jctc.6b00186) of double-stranded DNA simulations revealed  that `bsc1` gives similar results using either TIP3P or OPC water models; (c) TIP3P is significantly *cheaper* in terms of computer time than  OPC.

### **Ion parameters depend on water models** 

Since nucleic acids have a large number of negative charges distributed over the phosphate backbone, it is important to make an appropriate choice of  the ion parameters  representing the counterions  that are to be added to the model systems.  For typical monoatomic counterions ( $\ce{Na+}$, $\ce{K+}$, $\ce{Mg^{2+}}$, $\ce{Cl-}$, ...), several MM approaches have been pursued to reproduce both  their hydration shell structure and their hydration free energies, what may be  a challenging problem because ion···water interactions are relatively strong electrostatic forces modulated by the ionic size and *polarizability*.  

The simplest MM ion models incorporate only nonbonding parameters,  that is, a fixed point-charge combined with 12-6 LJ parameters. In this way, the ion potentials are   computationally efficient and  readily compatible with the AMBER/water models. For the monovalent ions ($\ce{Na+}$,$\ce{K+}$,..., $\ce{F-}$, $\ce{Cl-}$,...), a consistent set of 12-6 LJ parameters has been derived by  fitting solvation free energies,  radial distribution functions of water molecules, ion-water interaction energies and crystal lattice energies and lattice constants. These parameters, which constitute the **`JC` (Joung and Cheatham) family of ion parameters**,   have been optimized for each of the major water models (TIP3P, OPC, OPC3, TIP4P,...). Therefore, the ion parameters are *specific* sets that depend on the selection of the water model. 

**The JC  parameters for monovalent ions are routinely used in DNA simulations**.

The limitations of the two-parameter LJ potential become apparent in the case of  divalent and trivalent ions ($\ce{Mg^{2+}}$, $\ce{Zn^{2+}}$) for which it is very difficult to predict simultaneously  accurate hydration free energies and correct hydration shell structure.  It is possible then to select  specific parameterizations optimized to reproduce  either energetics or structural properties (but not both) as well as to select more flexible (but more *expensive*) 12-6-4 forms of the LJ potential (see the AMBER manual for further information).   


## **Periodic boundary conditions and long-range interactions**

### **Computer simulation of liquids: box tinkering** 
The general goal of the MD simulations is to characterize the structural and energetic properties of the modeled systems in aqueous solution. Considering that any liquid sample taken from a wet chemistry lab contains $N \sim 10^{23}$ atoms or molecules, some approximation is needed to keep  the value of $N$ within tractable bounds. The consensus approach is to prepare a model system containing $N$ atoms and place it formally in a container or cell, usually referred to as **solvent box** or simulation box.  The liquid (i.e. a condensed phase) is then described as a perfect periodic system consisting of an infinite number of replicas of the box generated by translational symmetry operations.  By imposing **periodic boundary conditions** (PBC) during a MD simulation,  only the atomic coordinates of the central box  need to be recorded and propagated. Under the PBC framework, when an atom leaves the central box by crossing the boundary, an image enters to replace it, and hence $N$  is conserved.

<img src="./PNG/PBC.png" width="640" height="480" style="display: block; margin: 0 auto">

**Fig 7** (a) A condensed phase described as a periodic arrangement of identical cells (replicas) (b) Illustration of periodic boundary conditions in 2D, showing the central squared box and its images (surrounding replicas) as adapted from (Schlick,  *Molecular Modeling and Simulation*).

Clearly, liquids are not periodic systems and the PBCs are artificial, but necessary to perform MD simulations of liquids. Thus, the PBCs allow production MD runs of systems with $ N \sim 10^5-10^6$ atoms on high performance computers  (up to $N \sim 10^8$ on supercomputers). Their goodness has been extensively validated by previous computational experience and comparison with experiment. However, to avoid artifacts (i.e. unrealistic behavior of biomolecules) that can be caused by PBCs, it is necessary to build sufficiently-large boxes. In practical terms, this means that **the *minimum distance* $d$ between any solute (DNA) atom and the box boundary  should be at least $d \geq 10\mathring{A}$**. If this condition is preserved during a certain  MD simulation, then non-bonding interactions between the central solute molecule  and its periodic images can be negligible. On one hand, short-range vdW attractions decay rapidly with intermolecular distance. On the other hand, long-range electrostatic interactions are largely dampened at interatomic distances $R\sim \geq 15-20 \mathring{A}$ thanks to the screening effect of water molecules and counterions. In this scenario, it turns out that MD simulations of **large solvent boxes**, in which solute-solute interactions play little or none role, are close to the *infinitely diluted* thermodynamic state.        

Besides  setting  the extent  of the solvent layer around  solute atoms (e.g. $d=15 \mathring{A}$), users must also select the **shape of the  box**. Of course, a regular shape is needed to completely fill the 3D space when the box is replicated. The most common option is to use cubic or rectangular (orthorombic) boxes. Other shapes that are possible are hexagonal and *truncated octahedral boxes*. Using the AMBER tools, only rectangular or truncated-octahedral boxes can be selected. 

For identical water thickness $d$ values, the box geometry affects the total number of atoms $N$ considered. Depending on the shape of solute, one or another box shape is more convenient. For example, a truncated octahedron may be more compact than a rectangular box for a globular protein, while a rectangular box would be more appropriate for elongated DNA systems. However, an important caveat is that the flexibility and structural stability of the solute must be also taken into account given that wide structural fluctuations and partial folding/unfolding events may result in strong PBC artifacts due to close interactions among image molecules (this may happen if, for example, a hairpin-like aptamer immersed in a rectangular box adopts an extended conformation).

#### **Guidelines for building a solvent box suitable for aptamers**
In the `APTAMD` protocol,  MD simulations are employed to perform extensive conformational sampling of the initial 3D models. Since the starting structures are poorly relaxed, they may experience ample structural transitions during the simulations as they access to more stable conformations. Localized fold/unfold events can thus be expected. Overall, conservative settings are required  to avoid  PBC artifacts:

* Use solvent layer with initial thickness of $d \geq 15 \mathring{A}$.
* Use truncated-octahedral boxes regardless of the initial shape of the aptamer molecule.

Certainly, a price is paid for these choices in terms of computer time. The resulting  boxes may contain a very high content of solvent atoms (even $\sim 99 \% $!), but they can accommodate wide structural motions of the solute that, in turn, can reveal the true dynamics of the isolated aptamer molecules. 

### **Overcoming the computational bottleneck due to long-range pairwise interactions**  

#### **Cutoff by necessity** 

Inspection of the AMBER potential formula points out immediately that  computational cost is determined by the pairwise interactions, which require evaluating a double sum with $\sim N^2$ terms. This results in a rapid, *quadratic* growth   $\mathcal{O} \left( N^2  \right)$ in computer time that compromises the applicability of the MM methods. To overcome the  $\mathcal{O}\left( N^2 \right)$ bottleneck,  the spherical cutoff method  assumes that  interactions between atoms are zero after some finite distance, the *cutoff distance* $r_c$. Then each atom 
interacts with all the atoms within a volume of $\frac{4}{3} \pi r_c^3$. If the mean particle density is $\rho$  the total number of pairwise interactions to be computed is $ \frac{4}{3}\pi r_c^3 \times \rho \times N  = \mathrm{Constant} \times N$. In this way, the computational cost is formally reduced to a linear scaling $\mathcal{O} \left( N \right)$ law. 

How to select the $r_c$ value? Basing on chemical intuition,  a $r_c \geq \sim 10 \mathcal{A}$ value should be considered so that van der Waals interactions, which decay rapidly with distance, would be nearly zero beyond $r_c$. In any case, some *truncation error* will exist associated to the long-range electrostatic forces. Note also that the value of $r_c$ is also restricted by the use of PBC, which requires  $r_c \leq a/2$ where $a$ is the smallest dimension of the solvent box. This condition implies the *minimum-image convention* so that each atom $i$ interacts only with the closest periodic image of the other $N−1$ atoms.

#### The Particle-Mesh-Ewald (PME) method: Quick and accurate computation of long-range effects ! 

We see in Figure 9 how the adoption of a cutoff method  drastically reduces the computational cost of simulations. However, typical cutoff values  introduce non-negligible truncation errors in the evaluation of long-range pairwise energies and forces. As  the upper limit to $r_c$ is also restricted by PBC,  the simple cutoff scheme produces poor simulations results for  highly charged systems such as the aptamer molecules.  

Fortunately, adoption of the PBC strategy paved the way to the inclusion of long-range effects in biomolecular simulations. For *periodic systems*, the so-called *Ewald lattice techniques* and, more particularly, the **Particle-Mesh-Ewald (PME) method**  evaluate the full electrostatic energy and forces  with a nearly linear-scaling cost  introducing only very small and strictly bounded numerical errors. The mathematical and technical details of the PME method are rather cumbersome, but a qualitative description can be helpful to better understand the settings of MD simulations. 

The *direct* calculation of the full electrostatic energy for periodic arrangement of solvent-box replicas demands us to calculate a very slowly convergent series:

$$ E_{elec} = \sum\limits_{i<j}^N \sum\limits_{\bf{n}} \frac{1}{4\pi\epsilon} \frac{q_i q_j}{\bf{r}_{ij}+\bf{n}} $$

where $N$ is the number of atoms in the solvent box and ${\bf{n} }=\left( n_x {\bf{a}}, n_y {\bf{b}}, n_z {\bf{c}} \right) $ are the *lattice vectors* that generate the infinite number of replicas of a rectangular box with $(a,b,c)$ dimensions. The Ewald method reformulates this series by adding and subtracting *Gaussian clouds*  centered at the point-charge positions. Mathematically, the Gaussians are defined by the width  parameter, $\alpha$.  In this way, the total charge density of the model system is:  

$$ \rho \left( \bf{r} \right) =  \sum\limits_i^{N} \left[ q_i \delta \left( \bf{r} - \bf{r}_i \right) - \frac{q_i \alpha^{(3/2)}}{\pi} e^{-\alpha^2\left(\bf{r}-\bf{r}_i \right)^2 } \right] + - \frac{q_i \alpha^{(3/2)}}{\pi} e^{-\alpha^2\left(\bf{r}-\bf{r}_i \right)^2 }$$

where the charge terms within the squared bracket would produce a rapidly decreasing potential.

**Fig 10**  1D schematic representation of the Ewald technique. Point charges are represented as $\delta$ functions (i.e. vertical bars).    

Although we avoid all mathematical and technical details, it is clear that the *Ewald trick* splits the total electrostatic energy into short- and long-range contributions. The great advantage  is that  each term can be  calculated by sums that converge rapidly and smoothly. On the one hand, the  short-range interaction between a point-charge $q_i$  and the  surrounding point charges + screening Gaussians, is computed accurately by a direct double sum using a cutoff $r_c$ scheme (i.e.  $\mathcal{O}\left( N\right)$ cost). On the other hand, the 
remaining long-range contribution involves only the continuous Gaussian functions, what allows to apply the *Fourier transform* technique that, in turn, results in a single sum of terms that depend on the lattice vectors of the *reciprocal space*. The computational effort is  now determined by  the Fourier transform operation, which has a cost of $\mathcal{O}\left(N^{3/2} \right)$, a significant reduction with respect to the quadratic scaling law.   

* **The Ewald method admits moderate cutoff radii of 8-10 $\mathring{A}$ that yield sufficiently-accurate results**.

**Fig 11**  1D schematic representation of the PME technique.    

The cost of the long-range  sum  can be further reduced to ($\mathcal{O}\left( N\log{N}  \right)$ ) using the particle-mesh approach (PME). This technique interpolates first all the point-charges over a regular mesh of grid points and, subsequently, the grid charges are replaced by regularly-distributed Gaussian functions that can be subject to a Fast Fourier transform (FFT). The computational speedup conveys some numerical error inherent in the mesh interpolation technique, which depends on the grid spacing $\Delta r_g$ and the spline interpolation order $n$ ( typical values are $\Delta r_g=0.5\mathring{A}$ and $n=3$). Note that the [FFT technique](https://doi.org/10.1021/acs.jchemed.4c01439) is much used in Spectroscopy for converting spectroscopic signals from the time domain  into (the reciprocal)  frequency domain (i.e $I(t) \rightarrow I(\nu)$ ). 

* **Modern PME implementations  select *automatically* the best parameters $\left( \alpha,\: \Delta r_g, \: n \right)$  that  minimize numerical errors while preserving  computational efficiency.**

#### **PME is the default method used by MD programs to calculate long-range effects**.     


### **Energy minimization with  Molecular Mechanics**

In molecular modeling,   local minima on the PES are of particular interest as they correspond to stable geometries  of the model system.  To identify those molecular geometries that correspond to minimium energy points, we use a *minimization algorithm*  that finds the geometrieswhere the  $E\left( {\bf{R}} \right) $ function satisfies the corresponding critical point conditions: 

$$ \frac{ \partial E\left( {\bf{R}} \right) }{\partial R_i } = 0 \:\:\:   \frac{ \partial^2 E\left( {\bf{R}} \right)}{\partial R_i^2 } >0 $$ 

MM methods give us with as analytical formulas to perform fast calculations of  the energy  function $E\left( {\bf{R}} \right) $  and the gradient vector ${\bf{g}} \left( {\bf{R}} \right)\equiv \nabla  E\left( {\bf{R}} \right) \equiv\left(  \frac{ \partial E\left( {\bf{R}} \right) }{\partial R_i } \right)_{i=1,N} $ .  The direction of the  gradient at any point $\bf{R}$ indicates where the closest minimum lies, and its magnitude measures the steepness of the local slope. Thus, the energy of the system can be decreased by moving each atom in response to the *force* acting on it. Energy minimization methods that implement these ideas are termed *first-order* or *derivative* methods that rely on the information of energy  and gradient evaluations.      

The **steepest descent** (SD) algorithm, which is frequently used in MM and is the simplest derivative method, performs the following actions: 
1. Select the starting point $k=1$ as the initial configuration.
2. Loop over a given number of *steps* ($k\leq k_{max}$) or until the gradient is below a convergence threshold  (${\bf{g_k}}\leq \varepsilon$)
    1. Calculate the gradient and the step direction towards the next point $k+1$ as $ {\bf{d}}_k= -\frac{ {\bf{g}}_{k}}{ \lvert {\bf{g}}_{k} }$
    2. Determine the step length $\alpha_k$ by  performing an *approximated* line search along ${\bf{d}_k}$  
    3. Take the step $ {\bf{R}}_{k+1} =  {\bf{R}}_{k} + \alpha_k {\bf{d}}_k $

As the number of atoms $N$ in model systems is very large,  the SD algorithms implemented in MM software seek to reduce the number of energy calculations by applying a polynomial interpolation technique  that performs an *inexact* line search to find the $\alpha_k$ value. An extension of the SD method is the popular **conjugate gradient** (CG) algorithm that takes steps along a direction ${\bf{d}}_k$ from point ${\bf{R}}_k$ that is *conjugated* (*orthogonal*) to the direction taken in the previous step ${\bf{d}}_{k-1}$, that is, the CG  strategy takes advantage of the energy and gradient evaluations at steps $k$ and $k-1$ to predict the new point $k$. 



* **Q7** The code cell contains a demo of the SD algorithm. The PES is modeled as a 2D analytical function   $E(x,y)=(x^{2}+y-11)^{2}+(x+y^{2}-7)^{2} $ that is used to test the performance of minimization algorithms.
  1. Modify the value of the initial point coordinates and try to find all critical points on the PES.
  2. Is the SD able to locate all the critical points?  

````python
# This code cell illustrates the SD algorithm by finding 
# the local minima of the Himmelblau model PES
import numpy as np
import matplotlib.pyplot as plt

# IF DESIRED CHANGE  INITIAL POINT, MAXSTEPS, EPS 
initial_point=np.array([0, 5]) # initial point
maxstep= 15 # max number of steps 
eps = 0.5   # convergence threshold value 

# Size of the 2D sample space (DO NOT CHANGE THESE VALUES)
X1max=X2max=5.0
X1min=X2min=-5.0

# Define the PES model function (Himmelblau) and its analytical gradient
def E(X):
    x, y = X[0], X[1]
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2
def gradient(X):
    x, y = X[0], X[1]
    df_dx = 2 * (x**2 + y - 11) * (2 * x) + 2 * (x + y**2 - 7)
    df_dy = 2 * (x**2 + y - 11) + 2 * (x + y**2 - 7) * (2 * y)
    return np.array([df_dx, df_dy])
    
# APPROXIMATED line search function  (parabolic interpolation). 
#                                  TEST THE INFLUENCE OF THE LINE SEARCH PARAMETERS (alpha_init, c1, rho)
def line_search_quadratic(X, grad, alpha_init=1.0, c1=1e-2, rho=0.5):
    """Finds step size alpha using ONLY the current point and gradient."""
    alpha = alpha_init
    E_X = E(X)
    # Using normalized negative gradient as direction
    grad_norm = np.linalg.norm(grad)
    direction = -grad / grad_norm
    grad_dot_d = -grad_norm
    # Backtrack until the Armijo condition is satisfied
    while E(X + alpha * direction) > E_X + c1 * alpha * grad_dot_d:
        E_alpha = E(X + alpha * direction)
        # Calculate the analytical minimum of the fitted parabola
        alpha_new = - (grad_dot_d * alpha**2) / (2.0 * (E_alpha - E_X - grad_dot_d * alpha))
        # Safeguard fallback
        if np.isnan(alpha_new) or alpha_new < 0.1 * alpha or alpha_new > 0.9 * alpha:
            alpha = rho * alpha
        else:
            alpha = alpha_new
        if alpha < 1e-6:
            break    
    return alpha, direction

# Optimization Loop
current_point = initial_point
traj_X1 = np.array([])
traj_X2 = np.array([])
for i in range(maxstep):
    grad = gradient(current_point)
    if np.linalg.norm(grad) < eps:
        break
    alpha, direction = line_search_quadratic(current_point, grad)
    current_point = current_point + alpha * direction
    traj_X1=np.append(traj_X1,current_point[0])
    traj_X2=np.append(traj_X2,current_point[1])

# Meshgrid generation 
x1_vals = np.linspace( X1min, X1max , 100)
x2_vals = np.linspace( X2min, X2max , 100)
X1_grid, X2_grid = np.meshgrid(x1_vals, x2_vals)
Z = E(np.array([X1_grid, X2_grid]))  

# Plot results
plt.contour(X1_grid, X2_grid, Z, levels=40, cmap='viridis')
plt.plot( traj_X1, traj_X2, '-or', markersize=4, label=f'SD steps={i}')
plt.plot(traj_X1[-1], traj_X2[-1], 'ob', markersize=7, label='Final step')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title(f'PES model function X1_final={traj_X1[-1]:f} X2_final={traj_X2[-1]:f}')
plt.legend()
plt.colorbar()
plt.grid(True)
plt.show()
````


Energy minimization is widely used in several MM procedures. In the context of MD simulations,  **energy minimization is
used to prepare the system in order to relieve any unfavorable interactions or steric clashes in the initial model of the system**. To fulfill this particular goal,  a *full* energy minimization until reaching a very low residual gradient (e.g., a root-mean-squared of gradient  $\rm{RMSG}\leq 10^{-4} \: \rm{kcal/( mol\:\mathring{A}})$) is not needed. Therefore, the MM programs take a fixed number of minimization steps ($\sim 1000-2000$) and/or adopt a moderate convergence threshold ($\rm{RMSG} \leq 10^{-2}$),  normally combining a few hundreds of SD steps followed by CG steps. Achieving a tight convergence threshold demands *second-order* minimization methods (not discussed here) that are computationally much more expensive.            

## **Molecular edition**

*Molecular edition* is the first stage in a standard workflow for MD simulations. Starting from  3D coordinates  of a single biomolecule or an aggregate of biomolecules, it is necessary to perform many quite different actions such as  applying chemical changes/fixes in the initial structure,  assigning  (or deriving)  MM atom types and parameters, creating a solvent box, etc. The end result of all the molecular edition tasks is an initial model system ready to be simulated.   

### Edition protocol for  DNA aptamers 
As expected, some of the  actions taken in a molecular edition process depend on  the chemical and/or structural properties of the biomolecular system. For our particular systems of interest, single ssDNA aptamer molecules, we propose to apply the following  protocol:

1. **Preparation of the initial PDB file** Select one PDB file containing the 3D coordinates of the RNA-equivalent model generated by `RNAComposer` or alike. Clean up the PDB file keeping just `ATOM` records of nucleic acid atoms. **Warning:** When starting from NMR multimodel structures, get just the coordinates of MODEL 1.  
2. **RNA$\longrightarrow$DNA virtual mutation** Using the `tleap` tool available in the AMBER software, mutate ribose sugars into deoxyribose and U nucleobases into Ts.
3. **DNA parameter assignment and partial relaxation**
   1. Select our favorite AMBER force field, `DNA.bsc1`, and let `tleap` to create new parameter (`.top`) and coordinate (`.crd`) files.    
   2. Using the `sander` program, perform a fast restrained relaxation of the sugar and bases to mitigate bad contacts or geometries.
4. **Building of the solvent box** Edit with `tleap` the previously relaxed system and load the  `water.tip3p` model.  Build a truncated-octahedral solvent box centered at the center-of-mass  of the aptamer. Save the coordinates of the hydrated system.
5. **Adding counterions**
   1. Select an adequate $\ce{NaCl(aq)}$ concentration (e.g. $[\ce{NaCl}]=0.150$ M). Calculate the number of  $\ce{Na+}$ and $\ce{Cl-}$ counterions to be added to the solvent box in consonance with the $[\ce{NaCl}]$ value.
   2. Edit the hydrated systems with  `tleap` to incorporate the necessary $\ce{Na+}$ and $\ce{Cl-}$ counterions selecting  the `ions_jc.tip3p` force field.  Save the coordinate and parameter files of the full hydrated system. 
   3. `tleap` places counterions basing on the electrostatic potential created by the solute atoms. A better option for highly-charged systems is to place the counterions at random positions  within the solvent box  (excluding the solute region). This task can be performed using the `cpptraj` program. 

Most of the actions enumerated in this protocol require complex manipulations of various data types that need to be performed by specialized software tools like `tleap`. However, usage of `tleap` or any other edition tool is not trivial because these tools request a  large variety of settings to be specified. The good news are: (a) this or similar molecular edition workflows can be implemented in a driver script; (b)  single aptamer molecules belong to the category of *easily edited* systems. This means that automated workflows are normally robust and rarely fail.
